```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# 1. 최종 데이터셋 로딩
df = pd.read_csv('../data/processed/ask1_master_final.csv')

# 2. RDKit 분자 객체 생성
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]

# 3. 계산할 기술자 목록 정의
descriptor_names = {
    "MW": Descriptors.MolWt,
    "LogP": Descriptors.MolLogP,
    "HBD": Descriptors.NumHDonors,
    "HBA": Descriptors.NumHAcceptors,
    "TPSA": Descriptors.TPSA,
    "RotatableBonds": Descriptors.NumRotatableBonds,
    "AromaticRings": Descriptors.NumAromaticRings
}

# 4. 각 분자에 대해 기술자 계산
descriptor_data = []
for mol in mols:
    if mol:
        descriptors = {name: func(mol) for name, func in descriptor_names.items()}
        descriptor_data.append(descriptors)
    else:
        # 분자 변환 실패 시 NaN으로 채움
        descriptor_data.append({name: None for name in descriptor_names.keys()})

# 5. 계산된 기술자를 기존 데이터프레임에 추가
desc_df = pd.DataFrame(descriptor_data)
df_with_descriptors = pd.concat([df, desc_df], axis=1)

# 결측치 확인 및 처리
df_with_descriptors.dropna(inplace=True)

print("물리화학적 피처 추가 완료:")
print(df_with_descriptors.head())
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors, Descriptors
from rdkit.Chem.Scaffolds import MurckoScaffold
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import warnings
from rdkit import RDLogger

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')

# --- 1. 데이터 및 피처 준비 ---
print("--- 1. 데이터 로딩 및 피처 생성 ---")

# 이전 단계에서 기술자를 추가한 데이터프레임을 사용합니다.
# df_with_descriptors 가 이미 메모리에 있다고 가정합니다. 
# 만약 없다면, 먼저 이전 셀을 실행해주세요.
df = df_with_descriptors.copy()

# 1a. ECFP 핑거프린트 생성
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
fp = [rdMolDescriptors.GetMorganFingerprintAsBitVect(m, radius=2, nBits=2048) for m in mols]
X_fp = np.array(fp)

# 1b. 물리화학적 기술자(Descriptors) 추출
descriptor_cols = ['MW', 'LogP', 'HBD', 'HBA', 'TPSA', 'RotatableBonds', 'AromaticRings']
X_desc = df[descriptor_cols].values

# 1c. 두 피처셋 결합
X_combined = np.hstack((X_fp, X_desc))
y = df['pIC50'].values

print(f"피처 준비 완료. 총 피처 수: {X_combined.shape[1]}개")

# --- 2. Scaffold 생성 ---
scaffolds = {}
for i, mol in enumerate(mols):
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
    if scaffold not in scaffolds:
        scaffolds[scaffold] = []
    scaffolds[scaffold].append(i)
scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)

# --- 3. 5-Fold 교차 검증 ---
print("\n--- 3. 5-Fold Scaffold 교차 검증 시작 (ECFP + Descriptors) ---")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
all_preds, all_trues, fold_scores = [], [], []

for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
    print(f"\n===== Fold {fold+1}/5 ===== ")
    train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
    test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]

    X_train, y_train = X_combined[train_idx], y[train_idx]
    X_test, y_test = X_combined[test_idx], y[test_idx]
    
    model = lgb.LGBMRegressor(random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
    all_trues.extend(y_test)
    all_preds.extend(y_pred)
    print(f"  -> Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")

# --- 4. 최종 성능 집계 및 기록 ---
print("\n--- 4. 최종 성능 집계 및 WandB 기록 ---")
scores_df = pd.DataFrame(fold_scores)
avg_rmse = scores_df['RMSE'].mean()
std_rmse = scores_df['RMSE'].std()
avg_r2 = scores_df['R2'].mean()
std_r2 = scores_df['R2'].std()

print("\n교차 검증 최종 결과 (ECFP + Descriptors):")
print(scores_df)
print(f"\n평균 RMSE: {avg_rmse:.4f} ± {std_rmse:.4f}")
print(f"평균 R²: {avg_r2:.4f} ± {std_r2:.4f}")

# wandb 실험 초기화
wandb.init(project="ask1_ic50_prediction", name="lgbm-ecfp-descriptors-scaffold-5fold")
wandb.log({ "Average RMSE": avg_rmse, "Std RMSE": std_rmse, "Average R2": avg_r2, "Std R2": std_r2 })

# --- 5. 전체 결과 시각화 ---
plt.figure(figsize=(8, 8))
sns.scatterplot(x=all_trues, y=all_preds, alpha=0.5)
plt.plot([min(all_trues), max(all_trues)], [min(all_trues), max(all_trues)], color='red', linestyle='--', linewidth=2)
plt.title('Actual vs. Predicted pIC50 (ECFP + Descriptors)', fontsize=16)
plt.xlabel('Actual pIC50 (All Folds)', fontsize=12)
plt.ylabel('Predicted pIC50 (All Folds)', fontsize=12)
plt.grid(True)

plot_path = '../outputs/figures/ecfp_desc_performance.png'
plt.savefig(plot_path, dpi=300)
wandb.log({"performance_plot": wandb.Image(plot_path)})
print(f"\n성능 플롯이 '{plot_path}'에 저장되었습니다.")

wandb.finish()
```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold

import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GATConv, global_mean_pool

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import warnings
from rdkit import RDLogger

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')


# --- 1. 데이터 로딩 및 그래프 변환 ---
print("--- 1. 데이터 로딩 및 그래프 변환 ---")
try:
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"데이터 로딩 완료: {len(df)} 행")
except FileNotFoundError:
    print("Error: '../data/processed/ask1_master_final.csv' 파일을 찾을 수 없습니다.")
    exit()

# SMILES를 PyG 그래프 데이터로 변환하는 함수
def smiles_to_graph(smiles, y):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: return None

    # 원자(노드) 특징 추출
    atom_features = []
    for atom in mol.GetAtoms():
        atom_features.append([
            atom.GetAtomicNum(), atom.GetDegree(), atom.GetFormalCharge(),
            atom.GetHybridization(), atom.GetIsAromatic()
        ])
    x = torch.tensor(atom_features, dtype=torch.float)

    # 결합(엣지) 정보 추출
    edge_indices = []
    for bond in mol.GetBonds():
        edge_indices.append([bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()])
    edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()
    
    data = Data(x=x, edge_index=edge_index, y=torch.tensor([y], dtype=torch.float))
    return data

# 전체 데이터셋을 그래프 리스트로 변환
data_list = [smiles_to_graph(s, y) for s, y in zip(df['smiles'], df['pIC50'])]
data_list = [d for d in data_list if d is not None]
print(f"그래프 변환 완료: {len(data_list)} 개")


# --- 2. GAT 모델 정의 ---
class GATModel(nn.Module):
    def __init__(self, num_node_features, hidden_channels=64):
        super(GATModel, self).__init__()
        self.conv1 = GATConv(num_node_features, hidden_channels, heads=8, dropout=0.2)
        self.conv2 = GATConv(hidden_channels * 8, hidden_channels, heads=1, dropout=0.2)
        self.out = nn.Linear(hidden_channels, 1)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index).relu()
        x = global_mean_pool(x, batch)
        x = self.out(x)
        return x

# --- 3. Scaffold 생성 ---
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
scaffolds = {}
for i, mol in enumerate(mols):
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
    if scaffold not in scaffolds: scaffolds[scaffold] = []
    scaffolds[scaffold].append(i)
scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)


# --- 4. 5-Fold 교차 검증 및 훈련 ---
print("\n--- 4. 5-Fold Scaffold 교차 검증 시작 (GAT) ---")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
all_preds, all_trues, fold_scores = [], [], []

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
    print(f"\n===== Fold {fold+1}/5 ===== ")
    train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
    test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]

    train_dataset = [data_list[i] for i in train_idx]
    test_dataset = [data_list[i] for i in test_idx]

    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
    
    model = GATModel(num_node_features=train_dataset[0].num_node_features).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.MSELoss()
    
    # 훈련 루프
    model.train()
    for epoch in range(300): # (!!! 100 -> 300 에포크로 변경 !!!)
        for batch in train_loader:
            batch = batch.to(device)
            optimizer.zero_grad()
            out = model(batch)
            loss = criterion(out, batch.y.view(-1, 1))
            loss.backward()
            optimizer.step()
        if (epoch + 1) % 50 == 0:
            print(f"  Epoch {epoch+1}/300 completed.")

    # 평가 루프
    model.eval()
    fold_preds = []
    with torch.no_grad():
        for batch in test_loader:
            batch = batch.to(device)
            pred = model(batch)
            fold_preds.extend(pred.cpu().numpy())
    
    y_test = [data.y.item() for data in test_dataset]
    y_pred = np.array(fold_preds).flatten()
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
    all_trues.extend(y_test)
    all_preds.extend(y_pred)
    print(f"  -> Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")

# --- 5. 최종 성능 집계 및 기록 ---
print("\n--- 5. 최종 성능 집계 및 WandB 기록 ---")
scores_df = pd.DataFrame(fold_scores)
avg_rmse, std_rmse = scores_df['RMSE'].mean(), scores_df['RMSE'].std()
avg_r2, std_r2 = scores_df['R2'].mean(), scores_df['R2'].std()

print("\n교차 검증 최종 결과 (GAT):")
print(scores_df)
print(f"\n평균 RMSE: {avg_rmse:.4f} ± {std_rmse:.4f}")
print(f"평균 R²: {avg_r2:.4f} ± {std_r2:.4f}")

wandb.init(project="ask1_ic50_prediction", name="gat-scaffold-5fold-300epochs")
wandb.log({ "Average RMSE": avg_rmse, "Std RMSE": std_rmse, "Average R2": avg_r2, "Std R2": std_r2 })

# --- 6. 전체 결과 시각화 ---
plt.figure(figsize=(8, 8))
sns.scatterplot(x=all_trues, y=all_preds, alpha=0.5)
plt.plot([min(all_trues), max(all_trues)], [min(all_trues), max(all_trues)], color='red', linestyle='--', linewidth=2)
plt.title('Actual vs. Predicted pIC50 (GAT, 300 Epochs)', fontsize=16)
plt.xlabel('Actual pIC50 (All Folds)', fontsize=12)
plt.ylabel('Predicted pIC50 (All Folds)', fontsize=12)
plt.grid(True)
plot_path = '../outputs/figures/gat_300epochs_scaffold_cv_performance.png'
plt.savefig(plot_path, dpi=300)
wandb.log({"performance_plot": wandb.Image(plot_path)})
print(f"\n성능 플롯이 '{plot_path}'에 저장되었습니다.")

wandb.finish()
```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold

import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GATConv, global_mean_pool

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import warnings
from rdkit import RDLogger

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')


# --- 1. 데이터 로딩 및 그래프 변환 ---
print("--- 1. 데이터 로딩 및 그래프 변환 ---")
try:
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"데이터 로딩 완료: {len(df)} 행")
except FileNotFoundError:
    print("Error: '../data/processed/ask1_master_final.csv' 파일을 찾을 수 없습니다.")
    exit()

# SMILES를 PyG 그래프 데이터로 변환하는 함수
def smiles_to_graph(smiles, y):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: return None

    # 원자(노드) 특징 추출
    atom_features = []
    for atom in mol.GetAtoms():
        atom_features.append([
            atom.GetAtomicNum(), atom.GetDegree(), atom.GetFormalCharge(),
            atom.GetHybridization(), atom.GetIsAromatic()
        ])
    x = torch.tensor(atom_features, dtype=torch.float)

    # 결합(엣지) 정보 추출
    edge_indices = []
    for bond in mol.GetBonds():
        edge_indices.append([bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()])
    edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()
    
    data = Data(x=x, edge_index=edge_index, y=torch.tensor([y], dtype=torch.float))
    return data

# 전체 데이터셋을 그래프 리스트로 변환
data_list = [smiles_to_graph(s, y) for s, y in zip(df['smiles'], df['pIC50'])]
data_list = [d for d in data_list if d is not None]
print(f"그래프 변환 완료: {len(data_list)} 개")


# --- 2. GAT 모델 정의 ---
class GATModel(nn.Module):
    def __init__(self, num_node_features, hidden_channels=64):
        super(GATModel, self).__init__()
        self.conv1 = GATConv(num_node_features, hidden_channels, heads=8, dropout=0.2)
        self.conv2 = GATConv(hidden_channels * 8, hidden_channels, heads=1, dropout=0.2)
        self.out = nn.Linear(hidden_channels, 1)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index).relu()
        x = global_mean_pool(x, batch)
        x = self.out(x)
        return x

# --- 3. Scaffold 생성 ---
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
scaffolds = {}
for i, mol in enumerate(mols):
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
    if scaffold not in scaffolds: scaffolds[scaffold] = []
    scaffolds[scaffold].append(i)
scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)


# --- 4. 5-Fold 교차 검증 및 훈련 ---
print("\n--- 4. 5-Fold Scaffold 교차 검증 시작 (GAT) ---")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
all_preds, all_trues, fold_scores = [], [], []

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
    print(f"\n===== Fold {fold+1}/5 ===== ")
    train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
    test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]

    train_dataset = [data_list[i] for i in train_idx]
    test_dataset = [data_list[i] for i in test_idx]

    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
    
    model = GATModel(num_node_features=train_dataset[0].num_node_features).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.MSELoss()
    
    # 훈련 루프
    model.train()
    for epoch in range(100):
        for batch in train_loader:
            batch = batch.to(device)
            optimizer.zero_grad()
            out = model(batch)
            loss = criterion(out, batch.y.view(-1, 1))
            loss.backward()
            optimizer.step()

    # 평가 루프
    model.eval()
    fold_preds = []
    with torch.no_grad():
        for batch in test_loader:
            batch = batch.to(device)
            pred = model(batch)
            fold_preds.extend(pred.cpu().numpy())
    
    y_test = [data.y.item() for data in test_dataset]
    y_pred = np.array(fold_preds).flatten()
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
    all_trues.extend(y_test)
    all_preds.extend(y_pred)
    print(f"  -> Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")

# --- 5. 최종 성능 집계 및 기록 ---
print("\n--- 5. 최종 성능 집계 및 WandB 기록 ---")
scores_df = pd.DataFrame(fold_scores)
avg_rmse, std_rmse = scores_df['RMSE'].mean(), scores_df['RMSE'].std()
avg_r2, std_r2 = scores_df['R2'].mean(), scores_df['R2'].std()

print("\n교차 검증 최종 결과 (GAT):")
print(scores_df)
print(f"\n평균 RMSE: {avg_rmse:.4f} ± {std_rmse:.4f}")
print(f"평균 R²: {avg_r2:.4f} ± {std_r2:.4f}")

wandb.init(project="ask1_ic50_prediction", name="gat-scaffold-5fold")
wandb.log({ "Average RMSE": avg_rmse, "Std RMSE": std_rmse, "Average R2": avg_r2, "Std R2": std_r2 })

# --- 6. 전체 결과 시각화 ---
plt.figure(figsize=(8, 8))
sns.scatterplot(x=all_trues, y=all_preds, alpha=0.5)
plt.plot([min(all_trues), max(all_trues)], [min(all_trues), max(all_trues)], color='red', linestyle='--', linewidth=2)
plt.title('Actual vs. Predicted pIC50 (GAT)', fontsize=16)
plt.xlabel('Actual pIC50 (All Folds)', fontsize=12)
plt.ylabel('Predicted pIC50 (All Folds)', fontsize=12)
plt.grid(True)
plot_path = '../outputs/figures/gat_scaffold_cv_performance.png'
plt.savefig(plot_path, dpi=300)
wandb.log({"performance_plot": wandb.Image(plot_path)})
print(f"\n성능 플롯이 '{plot_path}'에 저장되었습니다.")

wandb.finish()
```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GATConv, global_mean_pool

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import warnings
from rdkit import RDLogger
import optuna
import pickle
import os
from datetime import datetime

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')

# 필요한 디렉터리 생성
os.makedirs('../outputs/models/', exist_ok=True)
os.makedirs('../outputs/figures/', exist_ok=True)

# --- 1. 데이터 로딩 및 특징 통계 계산 ---
print("--- 1. 데이터 로딩 및 특징 통계 계산 ---")
try:
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"데이터 로딩 완료: {len(df)} 행")
except FileNotFoundError:
    print("Error: '../data/processed/ask1_master_final.csv' 파일을 찾을 수 없습니다.")
    exit()

# 전체 데이터에서 원자 특징 통계 미리 계산
print("원자 특징 통계 계산 중...")
all_atom_features = []
for smiles in df['smiles']:
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        for atom in mol.GetAtoms():
            all_atom_features.append([
                atom.GetAtomicNum(), 
                atom.GetDegree(), 
                atom.GetFormalCharge(),
                int(atom.GetHybridization()), 
                int(atom.GetIsAromatic())
            ])

# 특징별 통계 계산
all_atom_features = np.array(all_atom_features)
feature_mins = all_atom_features.min(axis=0)
feature_maxs = all_atom_features.max(axis=0)
feature_means = all_atom_features.mean(axis=0)
feature_stds = all_atom_features.std(axis=0)

print(f"원자 특징 통계:")
feature_names = ['AtomicNum', 'Degree', 'FormalCharge', 'Hybridization', 'IsAromatic']
for i, name in enumerate(feature_names):
    print(f"  {name}: min={feature_mins[i]:.2f}, max={feature_maxs[i]:.2f}, mean={feature_means[i]:.2f}, std={feature_stds[i]:.2f}")

# --- 2. 개선된 그래프 변환 함수 (정규화 포함) ---
def smiles_to_graph_normalized(smiles, y, normalization='minmax'):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: 
        return None

    # 원자(노드) 특징 추출
    atom_features = []
    for atom in mol.GetAtoms():
        # 하이브리드화를 원핫 인코딩으로 변환
        hybridization_onehot = [0, 0, 0, 0, 0]  # SP, SP2, SP3, SP3D, SP3D2
        hyb = int(atom.GetHybridization())
        if hyb < 5:
            hybridization_onehot[hyb] = 1
        
        atom_features.append([
            atom.GetAtomicNum(), 
            atom.GetDegree(), 
            atom.GetFormalCharge(),
            int(atom.GetIsAromatic())
        ] + hybridization_onehot)  # 하이브리드화 원핫 추가
    
    if len(atom_features) == 0:
        return None
        
    x = torch.tensor(atom_features, dtype=torch.float)

    # 특징 정규화
    if normalization == 'minmax':
        # Min-Max 정규화 (0-1 범위)
        x_min = torch.tensor([feature_mins[0], feature_mins[1], feature_mins[2], feature_mins[4]] + [0]*5, dtype=torch.float)
        x_max = torch.tensor([feature_maxs[0], feature_maxs[1], feature_maxs[2], feature_maxs[4]] + [1]*5, dtype=torch.float)
        x = (x - x_min) / (x_max - x_min + 1e-8)
    elif normalization == 'standard':
        # 표준화 (평균 0, 분산 1)
        x_mean = torch.tensor([feature_means[0], feature_means[1], feature_means[2], feature_means[4]] + [0.2]*5, dtype=torch.float)
        x_std = torch.tensor([feature_stds[0], feature_stds[1], feature_stds[2], feature_stds[4]] + [0.4]*5, dtype=torch.float)
        x = (x - x_mean) / (x_std + 1e-8)

    # 결합(엣지) 정보 추출
    edge_indices = []
    edge_attrs = []  # 결합 타입 정보도 추가
    for bond in mol.GetBonds():
        edge_indices.extend([
            [bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()],
            [bond.GetEndAtomIdx(), bond.GetBeginAtomIdx()]  # 무방향 그래프
        ])
        # 결합 타입 원핫 인코딩
        bond_type_onehot = [0, 0, 0, 0]  # SINGLE, DOUBLE, TRIPLE, AROMATIC
        bond_type = int(bond.GetBondType())
        if bond_type < 4:
            bond_type_onehot[bond_type-1] = 1
        edge_attrs.extend([bond_type_onehot, bond_type_onehot])  # 양방향
    
    if len(edge_indices) == 0:
        edge_index = torch.zeros((2, 0), dtype=torch.long)
        edge_attr = torch.zeros((0, 4), dtype=torch.float)
    else:
        edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()
        edge_attr = torch.tensor(edge_attrs, dtype=torch.float)
    
    data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, y=torch.tensor([y], dtype=torch.float))
    return data

# 전체 데이터셋을 그래프 리스트로 변환
print("\n--- 그래프 변환 중 (정규화 적용) ---")
data_list = []
for i, (s, y) in enumerate(zip(df['smiles'], df['pIC50'])):
    graph = smiles_to_graph_normalized(s, y, normalization='standard')
    if graph is not None:
        data_list.append(graph)
    if (i + 1) % 1000 == 0:
        print(f"  진행상황: {i+1}/{len(df)} 완료")

print(f"그래프 변환 완료: {len(data_list)} 개")
print(f"노드 특징 차원: {data_list[0].x.shape[1]}개")

# --- 3. 개선된 GAT 모델 정의 ---
class ImprovedGATModel(nn.Module):
    def __init__(self, num_node_features, hidden_channels=64, num_heads=8, dropout=0.2, num_layers=3, use_edge_attr=False):
        super(ImprovedGATModel, self).__init__()
        self.num_layers = num_layers
        self.dropout = dropout
        self.use_edge_attr = use_edge_attr
        
        # GAT 레이어들
        self.convs = nn.ModuleList()
        self.batch_norms = nn.ModuleList()
        
        # 첫 번째 GAT 레이어
        self.convs.append(GATConv(num_node_features, hidden_channels, heads=num_heads, dropout=dropout, edge_dim=4 if use_edge_attr else None))
        self.batch_norms.append(nn.BatchNorm1d(hidden_channels * num_heads))
        
        # 중간 GAT 레이어들
        for _ in range(num_layers - 2):
            self.convs.append(GATConv(hidden_channels * num_heads, hidden_channels, heads=num_heads, dropout=dropout, edge_dim=4 if use_edge_attr else None))
            self.batch_norms.append(nn.BatchNorm1d(hidden_channels * num_heads))
        
        # 마지막 GAT 레이어 (헤드 1개)
        if num_layers > 1:
            self.convs.append(GATConv(hidden_channels * num_heads, hidden_channels, heads=1, dropout=dropout, edge_dim=4 if use_edge_attr else None))
            self.batch_norms.append(nn.BatchNorm1d(hidden_channels))
        
        # 출력 레이어들 (더 복잡한 출력 구조)
        self.fc_layers = nn.ModuleList([
            nn.Linear(hidden_channels, hidden_channels // 2),
            nn.Linear(hidden_channels // 2, hidden_channels // 4),
            nn.Linear(hidden_channels // 4, 1)
        ])
        
        # 드롭아웃 레이어들
        self.dropout_layers = nn.ModuleList([nn.Dropout(dropout) for _ in range(3)])

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        edge_attr = data.edge_attr if self.use_edge_attr and hasattr(data, 'edge_attr') else None
        
        # GAT 레이어들 통과
        for i, (conv, bn) in enumerate(zip(self.convs, self.batch_norms)):
            if edge_attr is not None and self.use_edge_attr:
                x = conv(x, edge_index, edge_attr)
            else:
                x = conv(x, edge_index)
            
            x = bn(x)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
        
        # Global pooling
        x = global_mean_pool(x, batch)
        
        # FC 레이어들 통과
        for i, (fc, dropout_layer) in enumerate(zip(self.fc_layers, self.dropout_layers)):
            x = fc(x)
            if i < len(self.fc_layers) - 1:  # 마지막 레이어가 아닌 경우
                x = F.relu(x)
                x = dropout_layer(x)
        
        return x

# --- 4. Scaffold 생성 ---
print("\n--- Scaffold 생성 중... ---")
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
scaffolds = {}
for i, mol in enumerate(mols):
    if mol is not None:
        scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
        if scaffold not in scaffolds:
            scaffolds[scaffold] = []
        scaffolds[scaffold].append(i)

# 유효한 인덱스만 유지
valid_indices = set(range(len(data_list)))
filtered_scaffolds = {}
for scaffold, indices in scaffolds.items():
    filtered_indices = [i for i in indices if i in valid_indices]
    if filtered_indices:
        filtered_scaffolds[scaffold] = filtered_indices

scaffold_lists = sorted(filtered_scaffolds.values(), key=len, reverse=True)
print(f"Scaffold 생성 완료: {len(scaffold_lists)} 개 그룹")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# --- 5. 개선된 Optuna 목적 함수 ---
def improved_objective(trial):
    # 확장된 하이퍼파라미터 범위
    hidden_channels = trial.suggest_categorical('hidden_channels', [32, 64, 128, 256])
    num_heads = trial.suggest_categorical('num_heads', [4, 8, 16])
    dropout = trial.suggest_float('dropout', 0.1, 0.5)
    num_layers = trial.suggest_int('num_layers', 2, 5)  # 레이어 수 확장
    learning_rate = trial.suggest_float('learning_rate', 5e-5, 1e-1, log=True)  # 범위 확장
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64, 128])
    epochs = trial.suggest_int('epochs', 30, 100)  # 테스트용으로 적당한 범위
    use_edge_attr = trial.suggest_categorical('use_edge_attr', [True, False])
    weight_decay = trial.suggest_float('weight_decay', 1e-6, 1e-3, log=True)
    
    # K-Fold 교차 검증 (빠른 테스트를 위해 3-fold)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    fold_rmses = []
    
    for fold, (train_scaffold_idx, val_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
        # 데이터 분할
        train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
        val_idx = [idx for i in val_scaffold_idx for idx in scaffold_lists[i]]

        train_dataset = [data_list[i] for i in train_idx]
        val_dataset = [data_list[i] for i in val_idx]

        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
        
        # 모델 초기화
        model = ImprovedGATModel(
            num_node_features=train_dataset[0].x.shape[1],
            hidden_channels=hidden_channels,
            num_heads=num_heads,
            dropout=dropout,
            num_layers=num_layers,
            use_edge_attr=use_edge_attr
        ).to(device)
        
        optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)
        criterion = nn.MSELoss()
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10)
        
        # 조기 종료 설정
        best_val_loss = float('inf')
        patience = 15
        patience_counter = 0
        
        # 훈련 루프
        model.train()
        for epoch in range(epochs):
            epoch_loss = 0
            for batch in train_loader:
                batch = batch.to(device)
                optimizer.zero_grad()
                out = model(batch)
                loss = criterion(out, batch.y.view(-1, 1))
                loss.backward()
                
                # 그래디언트 클리핑
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                
                optimizer.step()
                epoch_loss += loss.item()
            
            # 검증 (매 3 에포크마다)
            if (epoch + 1) % 3 == 0:
                model.eval()
                val_loss = 0
                with torch.no_grad():
                    for batch in val_loader:
                        batch = batch.to(device)
                        pred = model(batch)
                        val_loss += criterion(pred, batch.y.view(-1, 1)).item()
                
                val_loss /= len(val_loader)
                scheduler.step(val_loss)
                
                # 조기 종료 체크
                if val_loss < best_val_loss:
                    best_val_loss = val_loss
                    patience_counter = 0
                else:
                    patience_counter += 1
                    if patience_counter >= patience:
                        break
                
                model.train()
        
        # 최종 평가
        model.eval()
        fold_preds = []
        with torch.no_grad():
            for batch in val_loader:
                batch = batch.to(device)
                pred = model(batch)
                fold_preds.extend(pred.cpu().numpy())
        
        y_val = [data.y.item() for data in val_dataset]
        y_pred = np.array(fold_preds).flatten()
        
        fold_rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        fold_rmses.append(fold_rmse)
        
        # 중간 결과 보고
        trial.report(fold_rmse, fold)
        if trial.should_prune():
            raise optuna.exceptions.TrialPruned()
    
    # 평균 RMSE 반환
    mean_rmse = np.mean(fold_rmses)
    print(f"Trial {trial.number}: RMSE={mean_rmse:.4f}, params={trial.params}")
    
    return mean_rmse

# --- 6. Optuna 튜닝 실행 (개선된 버전, 20번 시행) ---
print(f"\n--- 6. 개선된 Optuna 튜닝 시작 (20번 시행) ---")
print(f"튜닝 시작: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Study 생성
study = optuna.create_study(
    direction='minimize',
    pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=2),
    study_name=f"ImprovedGAT_ASK1_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
)

# WandB 초기화
wandb.init(project="ask1_improved_gat", name=f"improved-gat-20trials-{datetime.now().strftime('%m%d_%H%M')}")

# 20번 시행 실행
try:
    study.optimize(improved_objective, n_trials=20)
    
    # WandB 로깅
    for i, trial in enumerate(study.trials):
        if trial.state == optuna.trial.TrialState.COMPLETE:
            wandb.log({
                "trial": i,
                "rmse": trial.value,
                "params": trial.params
            })

except KeyboardInterrupt:
    print("\n사용자에 의해 중단되었습니다.")

# --- 7. 결과 정리 ---
print(f"\n--- 7. 개선된 튜닝 결과 정리 ---")
print(f"총 시행 횟수: {len(study.trials)}")
print(f"완료된 시행: {len([t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE])}")

if study.best_trial:
    print(f"\n최적 RMSE: {study.best_value:.4f}")
    print("최적 하이퍼파라미터:")
    for key, value in study.best_params.items():
        print(f"  {key}: {value}")
    
    # WandB에 최적 결과 로깅
    wandb.log({
        "best_rmse": study.best_value,
        "best_params": study.best_params,
        "total_trials": len(study.trials)
    })

    # --- 8. 최적 모델로 최종 평가 ---
    print(f"\n--- 8. 최적 모델로 최종 평가 ---")
    
    best_params = study.best_params
    
    # 3-fold 교차 검증
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    all_preds, all_trues, fold_scores = [], [], []
    
    for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
        print(f"Final Fold {fold+1}/3")
        train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
        test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]

        train_dataset = [data_list[i] for i in train_idx]
        test_dataset = [data_list[i] for i in test_idx]

        train_loader = DataLoader(train_dataset, batch_size=best_params['batch_size'], shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=best_params['batch_size'], shuffle=False)
        
        # 최적 모델 생성
        model = ImprovedGATModel(
            num_node_features=train_dataset[0].x.shape[1],
            hidden_channels=best_params['hidden_channels'],
            num_heads=best_params['num_heads'],
            dropout=best_params['dropout'],
            num_layers=best_params['num_layers'],
            use_edge_attr=best_params['use_edge_attr']
        ).to(device)
        
        optimizer = optim.Adam(model.parameters(), lr=best_params['learning_rate'], 
                              weight_decay=best_params['weight_decay'])
        criterion = nn.MSELoss()
        
        # 훈련
        model.train()
        for epoch in range(best_params['epochs']):
            for batch in train_loader:
                batch = batch.to(device)
                optimizer.zero_grad()
                out = model(batch)
                loss = criterion(out, batch.y.view(-1, 1))
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()

        # 평가
        model.eval()
        fold_preds = []
        with torch.no_grad():
            for batch in test_loader:
                batch = batch.to(device)
                pred = model(batch)
                fold_preds.extend(pred.cpu().numpy())
        
        y_test = [data.y.item() for data in test_dataset]
        y_pred = np.array(fold_preds).flatten()
        
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        
        fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
        all_trues.extend(y_test)
        all_preds.extend(y_pred)
        print(f"  Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")
    
    # 최종 성능 계산
    if fold_scores:
        scores_df = pd.DataFrame(fold_scores)
        avg_rmse = scores_df['RMSE'].mean()
        std_rmse = scores_df['RMSE'].std()
        avg_r2 = scores_df['R2'].mean()
        std_r2 = scores_df['R2'].std()
        
        print(f"\n개선된 GAT 최종 성능:")
        print(scores_df)
        print(f"평균 RMSE: {avg_rmse:.4f} ± {std_rmse:.4f}")
        print(f"평균 R²: {avg_r2:.4f} ± {std_r2:.4f}")
        
        # WandB에 최종 결과 로깅
        wandb.log({
            "final_avg_rmse": avg_rmse,
            "final_std_rmse": std_rmse,
            "final_avg_r2": avg_r2,
            "final_std_r2": std_r2,
            "improvement": "significant" if avg_r2 > 0.5 else "moderate"
        })
        
        # 시각화 (개선 확인)
        if len(all_trues) > 0 and len(all_preds) > 0:
            plt.figure(figsize=(12, 8))
            
            plt.subplot(1, 2, 1)
            plt.scatter(all_trues, all_preds, alpha=0.6, s=20)
            plt.plot([min(all_trues), max(all_trues)], [min(all_trues), max(all_trues)], 
                     color='red', linestyle='--', linewidth=2)
            plt.title('Improved GAT: Actual vs. Predicted pIC50', fontsize=14)
            plt.xlabel('Actual pIC50')
            plt.ylabel('Predicted pIC50')
            plt.grid(True)
            
            # 성능 지표 텍스트
            plt.text(0.05, 0.95, 
                     f'RMSE: {avg_rmse:.3f} ± {std_rmse:.3f}\n'
                     f'R²: {avg_r2:.3f} ± {std_r2:.3f}\n'
                     f'Trials: {len(study.trials)}', 
                     transform=plt.gca().transAxes, fontsize=10, verticalalignment='top',
                     bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
            
            # 잔차 플롯
            plt.subplot(1, 2, 2)
            residuals = np.array(all_trues) - np.array(all_preds)
            plt.scatter(all_preds, residuals, alpha=0.6, s=20)
            plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
            plt.title('Residuals Plot', fontsize=14)
            plt.xlabel('Predicted pIC50')
            plt.ylabel('Residuals (Actual - Predicted)')
            plt.grid(True)
            
            plt.tight_layout()
            plot_path = f'../outputs/figures/improved_gat_{datetime.now().strftime("%Y%m%d_%H%M")}.png'
            plt.savefig(plot_path, dpi=300, bbox_inches='tight')
            wandb.log({"improved_performance_plot": wandb.Image(plot_path)})
            plt.show()
            
            print(f"개선된 성능 플롯이 {plot_path}에 저장되었습니다.")

# Study 저장
study_path = f'../outputs/models/improved_gat_study_{datetime.now().strftime("%Y%m%d_%H%M%S")}.pkl'
with open(study_path, 'wb') as f:
    pickle.dump(study, f)
print(f"Study가 {study_path}에 저장되었습니다.")

print(f"\n=== 개선된 GAT 모델 튜닝 완료 ===")
print(f"종료 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

wandb.finish()
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import torch
from transformers import AutoTokenizer, AutoModel
from tqdm.auto import tqdm # 진행 상황 표시
import warnings
from rdkit import RDLogger

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')

# --- 1. 데이터 로딩 ---
print("--- 1. 최종 데이터 로딩 ---")
try:
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"데이터 로딩 완료: {len(df)} 행")
except FileNotFoundError:
    print("Error: '../data/processed/ask1_master_final.csv' 파일을 찾을 수 없습니다.")
    exit()

# --- 2. MolBERT 모델 및 토크나이저 로딩 ---
print("\n--- 2. MolBERT 모델 로딩 중 (처음 실행 시 시간이 걸릴 수 있습니다) ---")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_name = "seyonec/ChemBERTa-zinc-base-v1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model_bert = AutoModel.from_pretrained(model_name).to(device)
model_bert.eval() # 평가 모드로 설정
print(f"MolBERT 모델 로딩 완료. Device: {device}")

# --- 3. MolBERT 임베딩 생성 ---
print("\n--- 3. MolBERT 임베딩 생성 중 ---")
smiles_list = df['smiles'].tolist()
embeddings = []
batch_size = 32 # GPU 메모리에 맞춰 배치 크기 조절

with torch.no_grad():
    for i in tqdm(range(0, len(smiles_list), batch_size)):
        batch_smiles = smiles_list[i:i+batch_size]
        inputs = tokenizer(batch_smiles, return_tensors='pt', padding=True, truncation=True, max_length=128).to(device)
        outputs = model_bert(**inputs)
        # [CLS] 토큰의 임베딩을 사용 (분자 전체를 대표)
        batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
        embeddings.append(batch_embeddings)

X = np.vstack(embeddings)
y = df['pIC50'].values
print("임베딩 생성 완료.")
print(f"피처 행렬 크기: {X.shape}")

# --- 4. Scaffold 생성 ---
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
scaffolds = {}
for i, mol in enumerate(mols):
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
    if scaffold not in scaffolds: scaffolds[scaffold] = []
    scaffolds[scaffold].append(i)
scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)

# --- 5. 5-Fold 교차 검증 (LightGBM 사용) ---
print("\n--- 5. 5-Fold Scaffold 교차 검증 시작 (MolBERT + LightGBM) ---")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
all_preds, all_trues, fold_scores = [], [], []

for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
    print(f"\n===== Fold {fold+1}/5 ===== ")
    train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
    test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]

    X_train, y_train = X[train_idx], y[train_idx]
    X_test, y_test = X[test_idx], y[test_idx]
    
    model_lgbm = lgb.LGBMRegressor(random_state=42)
    model_lgbm.fit(X_train, y_train)
    
    y_pred = model_lgbm.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
    all_trues.extend(y_test)
    all_preds.extend(y_pred)
    print(f"  -> Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")

# --- 6. 최종 성능 집계 및 기록 ---
scores_df = pd.DataFrame(fold_scores)
avg_rmse, std_rmse = scores_df['RMSE'].mean(), scores_df['RMSE'].std()
avg_r2, std_r2 = scores_df['R2'].mean(), scores_df['R2'].std()

print("\n교차 검증 최종 결과 (MolBERT + LightGBM):")
print(scores_df)
print(f"\n평균 RMSE: {avg_rmse:.4f} ± {std_rmse:.4f}")
print(f"평균 R²: {avg_r2:.4f} ± {std_r2:.4f}")

wandb.init(project="ask1_ic50_prediction", name="lgbm-molbert-scaffold-5fold")
wandb.log({ "Average RMSE": avg_rmse, "Std RMSE": std_rmse, "Average R2": avg_r2, "Std R2": std_r2 })

# --- 7. 전체 결과 시각화 ---
plt.figure(figsize=(8, 8))
sns.scatterplot(x=all_trues, y=all_preds, alpha=0.5)
plt.plot([min(all_trues), max(all_trues)], [min(all_trues), max(all_trues)], color='red', linestyle='--', linewidth=2)
plt.title('Actual vs. Predicted pIC50 (MolBERT + LightGBM)', fontsize=16)
plt.xlabel('Actual pIC50 (All Folds)', fontsize=12)
plt.ylabel('Predicted pIC50 (All Folds)', fontsize=12)
plt.grid(True)
plot_path = '../outputs/figures/molbert_lgbm_performance.png'
plt.savefig(plot_path, dpi=300)
wandb.log({"performance_plot": wandb.Image(plot_path)})
print(f"\n성능 플롯이 '{plot_path}'에 저장되었습니다.")

wandb.finish()
```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold

import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GATConv, global_mean_pool

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import warnings
from rdkit import RDLogger
import os
from datetime import datetime

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')

# 필요한 디렉터리 생성
os.makedirs('../outputs/models/', exist_ok=True)
os.makedirs('../outputs/figures/', exist_ok=True)

# --- 1. 데이터 로딩 및 그래프 변환 ---
print("--- 1. 데이터 로딩 및 그래프 변환 ---")
try:
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"데이터 로딩 완료: {len(df)} 행")
except FileNotFoundError:
    print("Error: '../data/processed/ask1_master_final.csv' 파일을 찾을 수 없습니다.")
    exit()

# SMILES를 PyG 그래프 데이터로 변환하는 함수
def smiles_to_graph(smiles, y):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: return None

    # 원자(노드) 특징 추출
    atom_features = []
    for atom in mol.GetAtoms():
        atom_features.append([
            atom.GetAtomicNum(), atom.GetDegree(), atom.GetFormalCharge(),
            atom.GetHybridization(), atom.GetIsAromatic()
        ])
    x = torch.tensor(atom_features, dtype=torch.float)

    # 결합(엣지) 정보 추출
    edge_indices = []
    for bond in mol.GetBonds():
        edge_indices.append([bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()])
    
    if len(edge_indices) == 0:
        edge_index = torch.zeros((2, 0), dtype=torch.long)
    else:
        edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()
    
    data = Data(x=x, edge_index=edge_index, y=torch.tensor([y], dtype=torch.float))
    return data

# 전체 데이터셋을 그래프 리스트로 변환
print("그래프 변환 중...")
data_list = []
for i, (s, y) in enumerate(zip(df['smiles'], df['pIC50'])):
    graph = smiles_to_graph(s, y)
    if graph is not None:
        data_list.append(graph)
    if (i + 1) % 1000 == 0:
        print(f"  진행상황: {i+1}/{len(df)} 완료")

print(f"그래프 변환 완료: {len(data_list)} 개")

# --- 2. GAT 모델 정의 ---
class GATModel(nn.Module):
    def __init__(self, num_node_features, hidden_channels=64):
        super(GATModel, self).__init__()
        self.conv1 = GATConv(num_node_features, hidden_channels, heads=8, dropout=0.2)
        self.conv2 = GATConv(hidden_channels * 8, hidden_channels, heads=1, dropout=0.2)
        self.out = nn.Linear(hidden_channels, 1)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index).relu()
        x = global_mean_pool(x, batch)
        x = self.out(x)
        return x

# --- 3. Scaffold 생성 ---
print("--- Scaffold 생성 중... ---")
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
scaffolds = {}
for i, mol in enumerate(mols):
    if mol is not None:
        scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
        if scaffold not in scaffolds: 
            scaffolds[scaffold] = []
        scaffolds[scaffold].append(i)

# 유효한 인덱스만 유지
valid_indices = set(range(len(data_list)))
filtered_scaffolds = {}
for scaffold, indices in scaffolds.items():
    filtered_indices = [i for i in indices if i in valid_indices]
    if filtered_indices:
        filtered_scaffolds[scaffold] = filtered_indices

scaffold_lists = sorted(filtered_scaffolds.values(), key=len, reverse=True)
print(f"Scaffold 생성 완료: {len(scaffold_lists)} 개 그룹")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# --- 4. 10번 반복 실행 ---
print(f"\n--- 4. GAT 모델 10번 반복 실행 시작 ---")
print(f"시작 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# 전체 결과를 저장할 리스트들
all_run_results = []
all_run_scores = []

# WandB 초기화 (전체 실험용)
wandb.init(project="ask1_gat_10runs", name=f"gat-10runs-{datetime.now().strftime('%m%d_%H%M')}")

# 10번 반복
for run in range(10):
    print(f"\n{'='*60}")
    print(f"           RUN {run+1}/10 시작")
    print(f"{'='*60}")
    
    run_start_time = datetime.now()
    
    # 5-Fold 교차 검증
    kf = KFold(n_splits=5, shuffle=True, random_state=42+run)  # 각 run마다 다른 seed
    all_preds, all_trues, fold_scores = [], [], []
    
    for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
        print(f"\n--- Run {run+1}, Fold {fold+1}/5 ---")
        
        train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
        test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]

        train_dataset = [data_list[i] for i in train_idx]
        test_dataset = [data_list[i] for i in test_idx]

        train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
        
        # 모델 초기화 (각 fold마다 새로운 모델)
        model = GATModel(num_node_features=train_dataset[0].num_node_features).to(device)
        optimizer = optim.Adam(model.parameters(), lr=0.001)
        criterion = nn.MSELoss()
        
        # 훈련 루프
        model.train()
        for epoch in range(100):  # 100 에포크
            for batch in train_loader:
                batch = batch.to(device)
                optimizer.zero_grad()
                out = model(batch)
                loss = criterion(out, batch.y.view(-1, 1))
                loss.backward()
                optimizer.step()
            
            # 진행 상황 출력 (20 에포크마다)
            if (epoch + 1) % 20 == 0:
                print(f"  Epoch {epoch+1}/100 completed.")

        # 평가 루프
        model.eval()
        fold_preds = []
        with torch.no_grad():
            for batch in test_loader:
                batch = batch.to(device)
                pred = model(batch)
                fold_preds.extend(pred.cpu().numpy())
        
        y_test = [data.y.item() for data in test_dataset]
        y_pred = np.array(fold_preds).flatten()
        
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        
        fold_scores.append({'Run': run+1, 'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
        all_trues.extend(y_test)
        all_preds.extend(y_pred)
        print(f"  -> Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")

    # Run별 성능 계산
    run_scores_df = pd.DataFrame(fold_scores[-5:])  # 마지막 5개 fold만
    run_avg_rmse = run_scores_df['RMSE'].mean()
    run_std_rmse = run_scores_df['RMSE'].std()
    run_avg_r2 = run_scores_df['R2'].mean()
    run_std_r2 = run_scores_df['R2'].std()
    
    run_end_time = datetime.now()
    run_duration = (run_end_time - run_start_time).total_seconds() / 60  # 분 단위
    
    # Run 결과 저장
    run_result = {
        'run': run+1,
        'avg_rmse': run_avg_rmse,
        'std_rmse': run_std_rmse,
        'avg_r2': run_avg_r2,
        'std_r2': run_std_r2,
        'duration_minutes': run_duration,
        'all_preds': all_preds.copy(),
        'all_trues': all_trues.copy()
    }
    all_run_results.append(run_result)
    all_run_scores.extend(fold_scores[-5:])  # 마지막 5개 fold 추가
    
    print(f"\n--- Run {run+1} 완료 ---")
    print(f"평균 RMSE: {run_avg_rmse:.4f} ± {run_std_rmse:.4f}")
    print(f"평균 R²: {run_avg_r2:.4f} ± {run_std_r2:.4f}")
    print(f"소요 시간: {run_duration:.1f}분")
    
    # WandB에 Run별 결과 로깅
    wandb.log({
        f"run_{run+1}_avg_rmse": run_avg_rmse,
        f"run_{run+1}_avg_r2": run_avg_r2,
        f"run_{run+1}_duration": run_duration,
        "completed_runs": run+1
    })

# --- 5. 전체 10번 실행 결과 분석 ---
print(f"\n{'='*60}")
print("           전체 10번 실행 결과 분석")
print(f"{'='*60}")

# 전체 성능 데이터프레임 생성
all_scores_df = pd.DataFrame(all_run_scores)
print("\n모든 실행의 Fold별 결과:")
print(all_scores_df)

# Run별 평균 성능
run_summary = []
for result in all_run_results:
    run_summary.append({
        'Run': result['run'],
        'Avg_RMSE': result['avg_rmse'],
        'Avg_R2': result['avg_r2'],
        'Duration_min': result['duration_minutes']
    })

run_summary_df = pd.DataFrame(run_summary)
print(f"\nRun별 평균 성능:")
print(run_summary_df)

# 전체 통계
overall_rmse_mean = run_summary_df['Avg_RMSE'].mean()
overall_rmse_std = run_summary_df['Avg_RMSE'].std()
overall_r2_mean = run_summary_df['Avg_R2'].mean()
overall_r2_std = run_summary_df['Avg_R2'].std()
total_duration = run_summary_df['Duration_min'].sum()

print(f"\n🎯 전체 10번 실행 최종 결과:")
print(f"   평균 RMSE: {overall_rmse_mean:.4f} ± {overall_rmse_std:.4f}")
print(f"   평균 R²:   {overall_r2_mean:.4f} ± {overall_r2_std:.4f}")
print(f"   총 소요시간: {total_duration:.1f}분 ({total_duration/60:.1f}시간)")

# 최고/최저 성능
best_run = run_summary_df.loc[run_summary_df['Avg_RMSE'].idxmin()]
worst_run = run_summary_df.loc[run_summary_df['Avg_RMSE'].idxmax()]

print(f"\n📈 최고 성능: Run {best_run['Run']} (RMSE: {best_run['Avg_RMSE']:.4f}, R²: {best_run['Avg_R2']:.4f})")
print(f"📉 최저 성능: Run {worst_run['Run']} (RMSE: {worst_run['Avg_RMSE']:.4f}, R²: {worst_run['Avg_R2']:.4f})")

# WandB에 최종 결과 로깅
wandb.log({
    "final_overall_rmse_mean": overall_rmse_mean,
    "final_overall_rmse_std": overall_rmse_std,
    "final_overall_r2_mean": overall_r2_mean,
    "final_overall_r2_std": overall_r2_std,
    "total_duration_hours": total_duration/60,
    "best_run_number": int(best_run['Run']),
    "best_run_rmse": best_run['Avg_RMSE'],
    "worst_run_rmse": worst_run['Avg_RMSE']
})

# --- 6. 결과 시각화 ---
print(f"\n--- 6. 결과 시각화 생성 중... ---")

# 1) Run별 성능 변화 플롯
plt.figure(figsize=(15, 10))

# 서브플롯 1: RMSE 변화
plt.subplot(2, 3, 1)
plt.plot(run_summary_df['Run'], run_summary_df['Avg_RMSE'], 'bo-', linewidth=2, markersize=8)
plt.fill_between(run_summary_df['Run'], 
                 run_summary_df['Avg_RMSE'] - overall_rmse_std, 
                 run_summary_df['Avg_RMSE'] + overall_rmse_std, 
                 alpha=0.3)
plt.title('RMSE across 10 Runs', fontsize=12)
plt.xlabel('Run Number')
plt.ylabel('Average RMSE')
plt.grid(True)

# 서브플롯 2: R² 변화
plt.subplot(2, 3, 2)
plt.plot(run_summary_df['Run'], run_summary_df['Avg_R2'], 'ro-', linewidth=2, markersize=8)
plt.fill_between(run_summary_df['Run'], 
                 run_summary_df['Avg_R2'] - overall_r2_std, 
                 run_summary_df['Avg_R2'] + overall_r2_std, 
                 alpha=0.3)
plt.title('R² across 10 Runs', fontsize=12)
plt.xlabel('Run Number')
plt.ylabel('Average R²')
plt.grid(True)

# 서브플롯 3: 성능 분포 (박스플롯)
plt.subplot(2, 3, 3)
plt.boxplot([run_summary_df['Avg_RMSE'], run_summary_df['Avg_R2']], 
            labels=['RMSE', 'R²'])
plt.title('Performance Distribution', fontsize=12)
plt.ylabel('Value')

# 서브플롯 4: 최고 성능 Run의 예측 결과
best_run_idx = int(best_run['Run']) - 1
best_preds = all_run_results[best_run_idx]['all_preds']
best_trues = all_run_results[best_run_idx]['all_trues']

plt.subplot(2, 3, 4)
plt.scatter(best_trues, best_preds, alpha=0.6, s=20)
plt.plot([min(best_trues), max(best_trues)], [min(best_trues), max(best_trues)], 
         'r--', linewidth=2)
plt.title(f'Best Run {int(best_run["Run"])} Predictions', fontsize=12)
plt.xlabel('Actual pIC50')
plt.ylabel('Predicted pIC50')
plt.grid(True)

# 서브플롯 5: 실행 시간
plt.subplot(2, 3, 5)
plt.bar(run_summary_df['Run'], run_summary_df['Duration_min'])
plt.title('Runtime per Run', fontsize=12)
plt.xlabel('Run Number')
plt.ylabel('Duration (minutes)')

# 서브플롯 6: 전체 성능 히스토그램
plt.subplot(2, 3, 6)
plt.hist(all_scores_df['RMSE'], bins=20, alpha=0.7, color='blue', label='RMSE')
plt.hist(all_scores_df['R2'], bins=20, alpha=0.7, color='red', label='R²')
plt.title('All Fold Performance Distribution', fontsize=12)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plot_path = f'../outputs/figures/gat_10runs_analysis_{datetime.now().strftime("%Y%m%d_%H%M")}.png'
plt.savefig(plot_path, dpi=300, bbox_inches='tight')
wandb.log({"ten_runs_analysis": wandb.Image(plot_path)})
plt.show()

print(f"분석 플롯이 {plot_path}에 저장되었습니다.")

# --- 7. 결과 CSV 저장 ---
# Run별 요약 저장
run_summary_path = f'../outputs/models/gat_10runs_summary_{datetime.now().strftime("%Y%m%d_%H%M")}.csv'
run_summary_df.to_csv(run_summary_path, index=False)

# 전체 fold 결과 저장
all_results_path = f'../outputs/models/gat_10runs_all_folds_{datetime.now().strftime("%Y%m%d_%H%M")}.csv'
all_scores_df.to_csv(all_results_path, index=False)

print(f"\n결과가 다음 파일들에 저장되었습니다:")
print(f"- Run 요약: {run_summary_path}")
print(f"- 전체 결과: {all_results_path}")

print(f"\n🎉 GAT 모델 10번 반복 실행 완료!")
print(f"종료 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

wandb.finish()
```

```python
# =========================
# MolBERT Optuna Tuning (RoBERTa-safe) + Progress Bars + 통합 시각화
# =========================

# --- 0. 환경 변수: torch import 전에 ---
import os
os.makedirs('../models/', exist_ok=True)
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# --- 일반 라이브러리 ---
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
import pickle
import time # === 추가: 실행 시간 측정을 위해 ===

# --- RDKit ---
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold
from rdkit import RDLogger

# --- Torch / HF ---
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import RobertaConfig, RobertaModel

# --- 기타 ---
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score # === 추가: R2 스코어 ===
import optuna
import wandb
from tqdm.auto import tqdm

# === 추가: 시각화를 위한 라이브러리 ===
import matplotlib
#matplotlib.use("Agg") # 헤드리스 환경(서버 등)에서 실행 가능하도록 설정
import matplotlib.pyplot as plt

# --- 설정 ---
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')
optuna.logging.set_verbosity(optuna.logging.WARNING)

NUM_TRIALS = 20
SHOW_BATCH_BAR = False

# ===================================================================
# 1-3. 데이터 로딩, 토크나이저, 데이터셋, 모델 클래스 (이전과 동일)
# ===================================================================
print("--- 1. 데이터 로딩 ---")
df = pd.read_csv('../data/processed/ask1_master_final.csv')
print(f"데이터 로딩 완료: {len(df)} 행")

class SMILESTokenizer:
    def __init__(self, max_length=512):
        self.vocab = {
            '<unk>': 0, '<pad>': 1, '<mask>': 2, '<cls>': 3, '<sep>': 4,
            'C': 5, 'N': 6, 'O': 7, 'S': 8, 'P': 9, 'F': 10, 'Cl': 11, 'Br': 12, 'I': 13,
            'c': 14, 'n': 15, 'o': 16, 's': 17, 'p': 18, '(': 19, ')': 20, '[': 21,
            ']': 22, '=': 23, '#': 24, '-': 25, '+': 26, '1': 27, '2': 28, '3': 29,
            '4': 30, '5': 31, '6': 32, '7': 33, '8': 34, '9': 35, '@': 36, '/': 37,
            '\\': 38, '.': 39, ':': 40, '%': 41
        }
        self.max_length = max_length
        self.pad_token_id = 1
        self.unk_token_id = 0
        self.mask_token_id = 2
        self.cls_token_id = 3
        self.sep_token_id = 4
    def tokenize(self, smiles: str):
        tokens, i = [], 0
        while i < len(smiles):
            if i < len(smiles) - 1 and smiles[i:i+2] in self.vocab:
                tokens.append(smiles[i:i+2]); i += 2
            elif smiles[i] in self.vocab:
                tokens.append(smiles[i]); i += 1
            else:
                tokens.append('<unk>'); i += 1
        return tokens
    def encode(self, smiles: str):
        toks = self.tokenize(smiles)
        ids = [self.cls_token_id] + [self.vocab.get(t, self.unk_token_id) for t in toks[:self.max_length-2]] + [self.sep_token_id]
        if len(ids) < self.max_length:
            ids = ids + [self.pad_token_id] * (self.max_length - len(ids))
        else:
            ids = ids[:self.max_length]
        input_ids = torch.tensor(ids, dtype=torch.long)
        attention_mask = (input_ids != self.pad_token_id).long()
        return {'input_ids': input_ids, 'attention_mask': attention_mask}

class SMILESDataset(Dataset):
    def __init__(self, smiles_list, labels, tokenizer: SMILESTokenizer):
        self.smiles_list = smiles_list
        self.labels = labels
        self.tokenizer = tokenizer
    def __len__(self): return len(self.smiles_list)
    def __getitem__(self, idx):
        enc = self.tokenizer.encode(str(self.smiles_list[idx]))
        return {'input_ids': enc['input_ids'], 'attention_mask': enc['attention_mask'], 'labels': torch.tensor(float(self.labels[idx]), dtype=torch.float)}

class MolBERTRegressor(nn.Module):
    def __init__(self, config: RobertaConfig):
        super().__init__()
        self.roberta = RobertaModel(config)
        self.regressor = nn.Linear(config.hidden_size, 1)
    def forward(self, input_ids, attention_mask):
        out = self.roberta(input_ids=input_ids, attention_mask=attention_mask)
        cls = out.last_hidden_state[:, 0, :]
        return self.regressor(cls)

# --- 4. Scaffold split 및 유틸리티 함수 (이전과 동일) ---
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
scaffolds = {}
for i, mol in enumerate(mols):
    if mol:
        scaf = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
        scaffolds.setdefault(scaf, []).append(i)
scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

def make_collate_fn(vocab_size: int):
    def collate_fn(batch):
        ids = torch.stack([b['input_ids'] for b in batch], 0)
        attn = torch.stack([b['attention_mask'] for b in batch], 0)
        labels = torch.stack([b['labels'] for b in batch], 0)
        return {'input_ids': ids, 'attention_mask': attn, 'labels': labels}
    return collate_fn

def safe_cuda_cleanup():
    try:
        if torch.cuda.is_available(): torch.cuda.empty_cache()
    except Exception: pass

# --- 5. Optuna 목적 함수 ---
def objective(trial):
    num_heads = trial.suggest_categorical('num_attention_heads', [4, 8, 12])
    head_dim = trial.suggest_categorical('head_dim', [32, 64])
    hidden_size = num_heads * head_dim
    inter_mul = trial.suggest_categorical('intermediate_multiplier', [2, 4])
    inter_size = hidden_size * inter_mul
    
    params = {
        'num_hidden_layers': trial.suggest_int('num_hidden_layers', 4, 10),
        'hidden_dropout_prob': trial.suggest_float('hidden_dropout_prob', 0.1, 0.3),
        'attention_probs_dropout_prob': trial.suggest_float('attention_probs_dropout_prob', 0.1, 0.3),
        'learning_rate': trial.suggest_float('learning_rate', 1e-5, 1e-4, log=True),
        'batch_size': trial.suggest_categorical('batch_size', [16, 32]),
        'epochs': trial.suggest_int('epochs', 10, 30),
        'weight_decay': trial.suggest_float('weight_decay', 1e-6, 1e-4, log=True),
    }

    max_length = 256  # 기본값으로 설정
    tok_cfg = SMILESTokenizer(max_length=max_length)
    vocab_size = len(tok_cfg.vocab)
    pos_emb = max_length + 8

    config = RobertaConfig(
        vocab_size=vocab_size, pad_token_id=tok_cfg.pad_token_id, bos_token_id=tok_cfg.cls_token_id,
        eos_token_id=tok_cfg.sep_token_id, type_vocab_size=1, max_position_embeddings=pos_emb,
        hidden_size=hidden_size, num_attention_heads=num_heads, intermediate_size=inter_size,
        num_hidden_layers=params['num_hidden_layers'], hidden_dropout_prob=params['hidden_dropout_prob'],
        attention_probs_dropout_prob=params['attention_probs_dropout_prob'],
    )

    kf = KFold(n_splits=3, shuffle=True, random_state=42)
    splits = list(kf.split(scaffold_lists))
    fold_rmses = []
    collate_fn = make_collate_fn(vocab_size)
    try:
        for fold_idx, (tr_idx, va_idx) in enumerate(splits, start=1):
            train_idx = [ix for i in tr_idx for ix in scaffold_lists[i]]
            val_idx   = [ix for i in va_idx for ix in scaffold_lists[i]]
            tok = SMILESTokenizer(max_length=max_length)
            train_ds = SMILESDataset(df.iloc[train_idx]['smiles'].values, df.iloc[train_idx]['pIC50'].values, tok)
            val_ds   = SMILESDataset(df.iloc[val_idx]['smiles'].values,   df.iloc[val_idx]['pIC50'].values,   tok)
            train_loader = DataLoader(train_ds, batch_size=params['batch_size'], shuffle=True, collate_fn=collate_fn)
            val_loader   = DataLoader(val_ds,   batch_size=params['batch_size'], shuffle=False, collate_fn=collate_fn)
            model = MolBERTRegressor(config).to(device)
            optimizer = optim.AdamW(model.parameters(), lr=params['learning_rate'], weight_decay=params['weight_decay'])
            criterion = nn.MSELoss()
            for epoch in range(1, params['epochs'] + 1):
                model.train()
                for batch in train_loader:
                    input_ids, attention_mask, labels = [b.to(device) for b in batch.values()]
                    optimizer.zero_grad(set_to_none=True)
                    preds = model(input_ids, attention_mask)
                    loss = criterion(preds, labels.unsqueeze(1))
                    loss.backward()
                    optimizer.step()
            model.eval()
            preds_all = []
            with torch.no_grad():
                for batch in val_loader:
                    input_ids, attention_mask = batch['input_ids'].to(device), batch['attention_mask'].to(device)
                    out = model(input_ids, attention_mask)
                    preds_all.extend(out.cpu().numpy().flatten())
            rmse = float(np.sqrt(mean_squared_error(df.iloc[val_idx]['pIC50'].values, preds_all)))
            fold_rmses.append(rmse)
            del model, optimizer, criterion
            safe_cuda_cleanup()
    except Exception as e:
        tqdm.write(f"[Trial Failed] {repr(e)}")
        safe_cuda_cleanup()
        return float('inf')
    mean_rmse = float(np.mean(fold_rmses)) if fold_rmses else float('inf')
    trial.set_user_attr('fold_rmses', fold_rmses)
    return mean_rmse

# === 수정된 평가 함수 ===
def evaluate_best_model(best_params):
    print("\n--- 최적 모델 재평가 및 결과 수집 ---")
    
    # max_length를 기본값으로 설정
    max_length = 256
    
    tok = SMILESTokenizer(max_length=max_length)
    vocab_size = len(tok.vocab)
    collate_fn = make_collate_fn(vocab_size)
    
    # 구조 제약 (objective 함수와 동일하게)
    num_heads = best_params['num_attention_heads']
    head_dim = best_params['head_dim']
    hidden_size = num_heads * head_dim
    inter_mul = best_params['intermediate_multiplier']
    inter_size = hidden_size * inter_mul
    
    pos_emb = max_length + 8
    
    config = RobertaConfig(
        vocab_size=vocab_size, 
        pad_token_id=tok.pad_token_id, 
        bos_token_id=tok.cls_token_id,
        eos_token_id=tok.sep_token_id, 
        type_vocab_size=1, 
        max_position_embeddings=pos_emb,
        hidden_size=hidden_size, 
        num_attention_heads=num_heads, 
        intermediate_size=inter_size,
        num_hidden_layers=best_params['num_hidden_layers'], 
        hidden_dropout_prob=best_params['hidden_dropout_prob'],
        attention_probs_dropout_prob=best_params['attention_probs_dropout_prob'],
    )
    
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    splits = list(kf.split(scaffold_lists))
    all_preds, all_trues, fold_scores = [], [], []
    
    try:
        for fold_idx, (tr_idx, va_idx) in enumerate(splits, start=1):
            print(f"재평가 Fold {fold_idx}/5")
            train_idx = [ix for i in tr_idx for ix in scaffold_lists[i]]
            test_idx = [ix for i in va_idx for ix in scaffold_lists[i]]
            
            tok = SMILESTokenizer(max_length=max_length)
            train_ds = SMILESDataset(df.iloc[train_idx]['smiles'].values, df.iloc[train_idx]['pIC50'].values, tok)
            test_ds = SMILESDataset(df.iloc[test_idx]['smiles'].values, df.iloc[test_idx]['pIC50'].values, tok)
            
            train_loader = DataLoader(train_ds, batch_size=best_params['batch_size'], shuffle=True, collate_fn=collate_fn)
            test_loader = DataLoader(test_ds, batch_size=best_params['batch_size'], shuffle=False, collate_fn=collate_fn)
            
            model = MolBERTRegressor(config).to(device)
            optimizer = optim.AdamW(model.parameters(), lr=best_params['learning_rate'], weight_decay=best_params['weight_decay'])
            criterion = nn.MSELoss()
            
            # 훈련
            for epoch in range(1, best_params['epochs'] + 1):
                model.train()
                for batch in train_loader:
                    input_ids, attention_mask, labels = [b.to(device) for b in batch.values()]
                    optimizer.zero_grad(set_to_none=True)
                    preds = model(input_ids, attention_mask)
                    loss = criterion(preds, labels.unsqueeze(1))
                    loss.backward()
                    optimizer.step()
            
            # 평가
            model.eval()
            fold_preds = []
            with torch.no_grad():
                for batch in test_loader:
                    input_ids, attention_mask = batch['input_ids'].to(device), batch['attention_mask'].to(device)
                    out = model(input_ids, attention_mask)
                    fold_preds.extend(out.cpu().numpy().flatten())
            
            y_test = df.iloc[test_idx]['pIC50'].values
            y_pred = np.array(fold_preds)
            
            rmse = float(np.sqrt(mean_squared_error(y_test, y_pred)))
            r2 = float(r2_score(y_test, y_pred))
            
            fold_scores.append({'Fold': fold_idx, 'RMSE': rmse, 'R2': r2})
            all_trues.extend(y_test.tolist())
            all_preds.extend(y_pred.tolist())
            
            print(f"  Fold {fold_idx} RMSE: {rmse:.4f}, R²: {r2:.4f}")
            
            del model, optimizer, criterion
            safe_cuda_cleanup()
    
    except Exception as e:
        print(f"재평가 실패: {e}")
        # 빈 결과 반환
        return [], [], pd.DataFrame({'Fold': [1], 'RMSE': [float('inf')], 'R2': [0.0]})
    
    scores_df = pd.DataFrame(fold_scores)
    return all_trues, all_preds, scores_df

# --- 6. Optuna 실행 ---
print(f"\n--- 6. MolBERT Optuna 튜닝 시작 (Trials: {NUM_TRIALS}) ---")
start_time = time.time()
study = optuna.create_study(direction='minimize', study_name="MolBERT_ASK1")
study.optimize(objective, n_trials=NUM_TRIALS, show_progress_bar=True)
total_time = time.time() - start_time

# --- 7. 결과 저장 ---
print(f"\n--- 7. MolBERT 튜닝 결과 정리 ---")
best_trial = study.best_trial
if best_trial:
    print(f"최적 RMSE: {best_trial.value:.4f}")
    print("최적 하이퍼파라미터:", best_trial.params)
    out_path = f'../models/molbert_optuna_study_{datetime.now().strftime("%Y%m%d_%H%M")}.pkl'
    with open(out_path, 'wb') as f: pickle.dump(study, f)
    print(f"Optuna study가 {out_path}에 저장되었습니다.")
else:
    print("성공적으로 완료된 Trial이 없습니다.")
    exit() # 시각화를 진행할 수 없으므로 종료

# ===================================================================
# 8. 통합 결과 시각화 (이전 코드를 여기에 통합)
# ===================================================================
print(f"\n--- 8. 결과 시각화 ---")

# 시각화에 필요한 데이터 수집
all_trues, all_preds, scores_df = evaluate_best_model(best_trial.params)
avg_r2 = scores_df['R2'].mean()
avg_rmse = scores_df['RMSE'].mean()

# Optuna Trial 상태 집계
completed_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE]
pruned_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.PRUNED]
failed_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.FAIL]

# 6분할 플롯 생성
fig, axes = plt.subplots(2, 3, figsize=(18, 11))
fig.suptitle('MolBERT-Optuna Tuning 종합 결과', fontsize=18, fontweight='bold')

# 1. 실제값 vs 예측값
axes[0, 0].scatter(all_trues, all_preds, alpha=0.5, s=15, edgecolors='k', linewidth=0.5)
lo = min(min(all_trues), min(all_preds))
hi = max(max(all_trues), max(all_preds))
axes[0, 0].plot([lo, hi], [lo, hi], 'r--', linewidth=2)
axes[0, 0].set_xlabel('Actual pIC50', fontsize=12)
axes[0, 0].set_ylabel('Predicted pIC50', fontsize=12)
axes[0, 0].set_title(f'Prediction Scatter Plot\nMean R² = {avg_r2:.3f}', fontsize=14)
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# 2. Optuna 최적화 과정
valid_trials = completed_trials
if valid_trials:
    trial_numbers = [t.number for t in valid_trials]
    trial_values = [t.value for t in valid_trials]
    axes[0, 1].plot(trial_numbers, trial_values, 'o-', color='dodgerblue', linewidth=1.5, markersize=5)
    axes[0, 1].axhline(y=best_trial.value, color='red', linestyle='--', linewidth=1.5, label=f'Best RMSE = {best_trial.value:.3f}')
    axes[0, 1].legend(fontsize=10)
axes[0, 1].set_xlabel('Trial Number', fontsize=12)
axes[0, 1].set_ylabel('Validation RMSE', fontsize=12)
axes[0, 1].set_title('Optuna Optimization History', fontsize=14)
axes[0, 1].grid(True, linestyle='--', alpha=0.6)

# 3. Fold별 R2 Score
x_pos = np.arange(len(scores_df))
axes[0, 2].bar(x_pos, scores_df['R2'], alpha=0.8, color='skyblue', edgecolor='black')
axes[0, 2].axhline(y=avg_r2, color='red', linestyle='--', linewidth=1.5, label=f'Mean R² = {avg_r2:.3f}')
axes[0, 2].set_xlabel('Fold', fontsize=12)
axes[0, 2].set_ylabel('R² Score', fontsize=12)
axes[0, 2].set_title('Performance by Fold', fontsize=14)
axes[0, 2].set_xticks(x_pos)
axes[0, 2].set_xticklabels(scores_df['Fold'].values)
axes[0, 2].legend(fontsize=10)
axes[0, 2].set_ylim(bottom=max(0, scores_df['R2'].min() - 0.1))
axes[0, 2].grid(True, axis='y', linestyle='--', alpha=0.6)

# 4. 잔차도
residuals = np.array(all_trues) - np.array(all_preds)
axes[1, 0].scatter(all_preds, residuals, alpha=0.5, s=15, edgecolors='k', linewidth=0.5)
axes[1, 0].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[1, 0].set_xlabel('Predicted pIC50', fontsize=12)
axes[1, 0].set_ylabel('Residuals (Actual - Predicted)', fontsize=12)
axes[1, 0].set_title('Residual Plot', fontsize=14)
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# 5. Trial 상태
trial_states = ['Completed', 'Pruned', 'Failed']
trial_counts = [len(completed_trials), len(pruned_trials), len(failed_trials)]
colors = ['#4CAF50', '#FFC107', '#F44336']
axes[1, 1].pie(trial_counts, labels=trial_states, autopct='%1.1f%%', startangle=90, colors=colors,
               wedgeprops={'edgecolor': 'white', 'linewidth': 1})
axes[1, 1].set_title(f'Trial Status (Total: {len(study.trials)})', fontsize=14)
axes[1, 1].axis('equal')

# 6. 결과 요약
summary_text = (
    f'MolBERT Tuning Summary\n\n'
    f'Best Validation RMSE: {best_trial.value:.4f}\n'
    f'Mean R² (on best model): {avg_r2:.4f}\n\n'
    f'Total Trials: {len(study.trials)}\n'
    f'Completed Trials: {len(completed_trials)}\n\n'
    f'Total Tuning Time: {total_time/60:.1f} min\n'
)
axes[1, 2].text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=12,
                bbox=dict(boxstyle='round,pad=0.5', fc='aliceblue', ec='k', lw=1, alpha=0.8))
axes[1, 2].axis('off')

# 전체 레이아웃 조정 및 저장
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # suptitle과 겹치지 않도록 조정

ts = datetime.now().strftime("%Y%m%d_%H%M")
fig_dir = f"../models/figs_{ts}"
os.makedirs(fig_dir, exist_ok=True)
plot_path = os.path.join(fig_dir, f'molbert_tuning_summary_{ts}.png')

try:
    plt.savefig(plot_path, dpi=300, bbox_inches='tight')
    print(f"📈 결과 플롯 저장: {plot_path}")
except Exception as e:
    print(f"⚠️ 그래프 저장 실패: {e}")

plt.show() # <-- 이 줄을 추가해주세요!
plt.close(fig) # 메모리 해제
print("[DONE] 시각화가 완료되었습니다.")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors, Descriptors, Lipinski
from rdkit.Chem.Scaffolds import MurckoScaffold
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import optuna
import pickle
import os
from datetime import datetime
import time

# --- 1. 저장된 Optuna Study 및 데이터 불러오기 ---
print("--- 1. 저장된 결과 로딩 ---")
study_path = '../models/optuna_study_final_lgbm.pkl' 
with open(study_path, 'rb') as f:
    study = pickle.load(f)

df = pd.read_csv('../data/processed/ask1_data_with_docking_scores.csv')

best_params = study.best_params
print(f"최적 RMSE (from study): {study.best_value:.4f}")

# --- 2. 최종 피처 준비 ---
print("\n--- 2. 최종 피처 준비 ---")
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
X_fp = np.array([rdMolDescriptors.GetMorganFingerprintAsBitVect(m, radius=2, nBits=2048) for m in mols])
descriptor_names = {
    "MolWt": Descriptors.MolWt, "LogP": Descriptors.MolLogP, "TPSA": Descriptors.TPSA,
    "HBD": Lipinski.NumHDonors, "HBA": Lipinski.NumHAcceptors, "RotatableBonds": Lipinski.NumRotatableBonds,
    "AromaticRings": Lipinski.NumAromaticRings, "FractionCSP3": Lipinski.FractionCSP3,
    "FormalCharge": Chem.rdmolops.GetFormalCharge, "RingCount": Lipinski.RingCount
}
desc_data = [{name: func(mol) for name, func in descriptor_names.items()} if mol else {name: None for name in descriptor_names.keys()} for mol in mols]
desc_df = pd.DataFrame(desc_data).fillna(0)
X_desc = desc_df.values
X_docking = df[['docking_score']].values
X_combined = np.hstack((X_fp, X_desc, X_docking))
y = df['pIC50'].values

# --- 3. Scaffold 생성 ---
scaffolds = {}
for i, mol in enumerate(mols):
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
    if scaffold not in scaffolds: scaffolds[scaffold] = []
    scaffolds[scaffold].append(i)
scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)

# --- 4. 최적 파라미터로 5-Fold 교차 검증 재실행 ---
print("\n--- 4. 최적 파라미터로 최종 평가 실행 ---")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
all_preds, all_trues, fold_scores = [], [], []

for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
    train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
    test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]
    X_train, y_train = X_combined[train_idx], y[train_idx]
    X_test, y_test = X_combined[test_idx], y[test_idx]
    
    model = lgb.LGBMRegressor(**best_params, random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
    all_trues.extend(y_test)
    all_preds.extend(y_pred)

scores_df = pd.DataFrame(fold_scores)
avg_rmse = scores_df['RMSE'].mean()
avg_r2 = scores_df['R2'].mean()
print("최종 평가 완료.")

# --- 5. 종합 대시보드 시각화 ---
print("\n--- 5. 결과 시각화 ---")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(f'Final Tuned LightGBM Analysis (Avg R² = {avg_r2:.3f})', fontsize=20)

# [0, 0] 예측 성능
axes[0, 0].scatter(all_trues, all_preds, alpha=0.5, s=15, edgecolor='k', linewidth=0.5)
lo, hi = min(min(all_trues), min(all_preds)), max(max(all_trues), max(all_preds))
axes[0, 0].plot([lo, hi], [lo, hi], 'r--', linewidth=2)
axes[0, 0].set_xlabel('Actual pIC50')
axes[0, 0].set_ylabel('Predicted pIC50')
axes[0, 0].set_title(f'Overall Performance')
axes[0, 0].grid(True, alpha=0.3)

# [0, 1] 튜닝 과정 (Optuna History Plot)
# Optuna의 시각화 함수를 Matplotlib subplot에 직접 삽입
optuna.visualization.matplotlib.plot_optimization_history(study, ax=axes[0, 1])
axes[0, 1].set_title('Optuna Optimization History')

# [0, 2] 폴드별 R² 성능
x_pos = np.arange(len(fold_scores))
axes[0, 2].bar(x_pos, scores_df['R2'], alpha=0.7, edgecolor='black', color='skyblue')
axes[0, 2].axhline(y=avg_r2, color='r', linestyle='--', linewidth=1.5, label=f'Mean = {avg_r2:.3f}')
axes[0, 2].set_xlabel('Fold')
axes[0, 2].set_ylabel('R² Score')
axes[0, 2].set_title('Performance by Fold')
axes[0, 2].set_xticks(x_pos)
axes[0, 2].set_xticklabels([f'F{i+1}' for i in range(len(fold_scores))])
axes[0, 2].legend()
axes[0, 2].grid(True, alpha=0.3)

# [1, 0] 잔차 분석
residuals = np.array(all_trues) - np.array(all_preds)
axes[1, 0].scatter(all_preds, residuals, alpha=0.5, s=15, edgecolor='k', linewidth=0.5)
axes[1, 0].axhline(y=0, color='r', linestyle='--', linewidth=2)
axes[1, 0].set_xlabel('Predicted pIC50')
axes[1, 0].set_ylabel('Residuals')
axes[1, 0].set_title('Residual Plot')
axes[1, 0].grid(True, alpha=0.3)

# [1, 1] 튜닝 상태 (Pie Chart)
completed = len([t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE])
pruned = len([t for t in study.trials if t.state == optuna.trial.TrialState.PRUNED])
failed = len([t for t in study.trials if t.state == optuna.trial.TrialState.FAIL])
axes[1, 1].pie([completed, pruned, failed], labels=['Completed', 'Pruned', 'Failed'], 
                autopct='%1.1f%%', startangle=90, colors=['#99ff99','#ffcc99','#ff9999'])
axes[1, 1].set_title(f'Trial Status (Total: {len(study.trials)})')

# [1, 2] 핵심 요약
summary_text = f'Final Model Summary\n\nBest RMSE (Optuna): {study.best_value:.4f}\nMean R² (CV): {avg_r2:.4f}\nMean RMSE (CV): {avg_rmse:.4f}'
axes[1, 2].text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=12,
                bbox=dict(boxstyle='round,pad=0.5', fc='aliceblue', ec='black', lw=1))
axes[1, 2].axis('off')

# 전체 레이아웃 조정 및 저장
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plot_path = f'../outputs/figures/final_dashboard_plot.png'
plt.savefig(plot_path, dpi=300)
print(f"\n📈 결과 대시보드 플롯 저장: {plot_path}")
plt.show()
```

```python
from rdkit import Chem
import numpy as np

# PDB 파일에서 리간드 원자만 선택하여 좌표를 계산합니다.
# PDB 파일은 docking 폴더에 있다고 가정합니다.
pdb_file = '../docking/6e2n_chainA.pdb'
mol = Chem.MolFromPDBFile(pdb_file, removeHs=False)

# PDB 파일을 열어 확인한 리간드의 잔기 이름(residue name)으로 수정해야 합니다.
LIGAND_RESIDUE_NAME = 'KK8' 

# 리간드 원자의 3D 좌표 추출
conformer = mol.GetConformer()
ligand_coords = []
for atom in mol.GetAtoms():
    pdb_info = atom.GetPDBResidueInfo()
    if pdb_info and pdb_info.GetResidueName().strip() == LIGAND_RESIDUE_NAME:
        pos = conformer.GetAtomPosition(atom.GetIdx())
        ligand_coords.append([pos.x, pos.y, pos.z])

ligand_coords = np.array(ligand_coords)

# 중심점과 크기 계산
center = ligand_coords.mean(axis=0)
max_coords = ligand_coords.max(axis=0)
min_coords = ligand_coords.min(axis=0)
size = max_coords - min_coords

# Gnina에 사용할 값 (여유 공간 5Å 추가)
gnina_center = center
gnina_size = size + 5.0

print(f"PDB 파일: {pdb_file}")
print(f"리간드 '{LIGAND_RESIDUE_NAME}'의 원자 {len(ligand_coords)}개 발견")
print("-" * 30)
print("Gnina에 사용할 좌표:")
print(f"  Center (x, y, z): {gnina_center[0]:.3f}, {gnina_center[1]:.3f}, {gnina_center[2]:.3f}")
print(f"  Size (x, y, z):   {gnina_size[0]:.3f}, {gnina_size[1]:.3f}, {gnina_size[2]:.3f}")
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
import os

# --- 1. 최종 데이터 및 저장 경로 설정 ---
df = pd.read_csv('../data/processed/ask1_master_final.csv')
ligand_dir = '../docking/ligands_sdf/'
os.makedirs(ligand_dir, exist_ok=True)

print(f"총 {len(df)}개의 리간드를 3D 구조로 변환합니다.")
print(f"저장 위치: {ligand_dir}")

# --- 2. SMILES를 3D SDF 파일로 변환 및 저장 ---
for i, row in df.iterrows():
    smiles = row['smiles']
    # 파일 이름으로 사용할 수 없는 문자를 '_'로 변경
    molecule_id = f"mol_{i}"
    
    # RDKit 분자 객체 생성
    mol = Chem.MolFromSmiles(smiles)
    
    if mol:
        # 3D 구조 생성을 위해 수소 추가
        mol_with_h = Chem.AddHs(mol)
        
        # 3D 구조 생성 및 에너지 최소화 (ETKDG 알고리즘 사용)
        AllChem.EmbedMolecule(mol_with_h, AllChem.ETKDG())
        AllChem.MMFFOptimizeMolecule(mol_with_h)
        
        # SDF 파일로 저장
        writer = Chem.SDWriter(os.path.join(ligand_dir, f"{molecule_id}.sdf"))
        writer.write(mol_with_h)
        writer.close()

print(f"\n리간드 변환 완료.")
```

```python
import pandas as pd
import os
import subprocess
from tqdm.auto import tqdm

# --- 1. 경로 및 좌표 설정 ---
print("--- 1. 경로 및 좌표 설정 ---")

project_root = '/home/nudge/Project/ASK1_portfolio/'
receptor_file = os.path.join(project_root, 'docking/6e2n_chainA_only.pdbqt')
ligand_dir = os.path.join(project_root, 'docking/ligands_sdf/')
output_dir = os.path.join(project_root, 'docking/outputs/')
gnina_executable = os.path.join(project_root, 'docking/gnina')

os.makedirs(output_dir, exist_ok=True)

# 이전에 계산한 도킹 박스 정보
center_x = -4.171
center_y = -25.526
center_z = 51.454
size_x = 14.082
size_y = 7.695
size_z = 18.814

# --- 2. 도킹 실행 및 결과 파싱 ---
print("\n--- 2. Gnina 도킹 실행 시작 (GPU 사용) ---")

results = []
ligand_files = sorted([f for f in os.listdir(ligand_dir) if f.endswith('.sdf')])

for ligand_file in tqdm(ligand_files, desc="Docking Ligands"):
    ligand_path = os.path.join(ligand_dir, ligand_file)
    molecule_id = ligand_file.split('.')[0]
    output_sdf_path = os.path.join(output_dir, f"{molecule_id}_out.sdf")
    
    # Gnina 실행 명령어 구성
    command = [
        gnina_executable,
        '-r', receptor_file,
        '-l', ligand_path,
        '--center_x', str(center_x),
        '--center_y', str(center_y),
        '--center_z', str(center_z),
        '--size_x', str(size_x),
        '--size_y', str(size_y),
        '--size_z', str(size_z),
        '-o', output_sdf_path,
        '--exhaustiveness', '8'
        # '--cpu' 옵션을 삭제하여 GPU를 자동으로 사용하도록 함
    ]
    
    # Gnina 실행
    try:
        result = subprocess.run(command, capture_output=True, text=True, check=True)
        
        # Gnina 출력에서 도킹 점수 파싱
        best_score = None
        for line in result.stdout.split('\n'):
            if "minimizedAffinity" in line:
                try:
                    score = float(line.split(':')[1].strip().replace(',', ''))
                    best_score = score
                    break
                except (IndexError, ValueError):
                    continue
        
        if best_score is not None:
            results.append({'molecule_id': molecule_id, 'docking_score': best_score})
            
    except subprocess.CalledProcessError as e:
        print(f"Error docking {ligand_file}: {e.stderr}")

# --- 3. 결과 저장 ---
print("\n--- 3. 도킹 결과 저장 ---")
docking_results_df = pd.DataFrame(results)
docking_results_df['original_index'] = docking_results_df['molecule_id'].str.replace('mol_', '').astype(int)

df = pd.read_csv('../data/processed/ask1_master_final.csv')
df_with_docking = pd.merge(df, docking_results_df, left_index=True, right_on='original_index')

final_output_path = '../data/processed/ask1_data_with_docking_scores.csv'
df_with_docking.to_csv(final_output_path, index=False)

print(f"\n도킹 완료. 최종 결과가 '{final_output_path}'에 저장되었습니다.")
print(df_with_docking[['smiles', 'pIC50', 'docking_score']].head())
```

```python
import pandas as pd
import os
from rdkit import Chem
from tqdm.auto import tqdm

# --- 1. 경로 설정 ---
project_root = '/home/nudge/Project/ASK1_portfolio/'
output_dir = os.path.join(project_root, 'docking/outputs/')
final_data_path = os.path.join(project_root, 'data/processed/')

# --- 2. 저장된 SDF 파일에서 도킹 점수 파싱 ---
print("--- 2. 저장된 도킹 결과 파일에서 점수 추출 시작 ---")

results = []
output_files = sorted([f for f in os.listdir(output_dir) if f.endswith('_out.sdf')])

for sdf_file in tqdm(output_files, desc="Parsing Scores"):
    molecule_id = sdf_file.replace('_out.sdf', '')
    sdf_path = os.path.join(output_dir, sdf_file)
    
    best_score = None
    try:
        # RDKit으로 SDF 파일 로드
        suppl = Chem.SDMolSupplier(sdf_path)
        # SDF 파일에는 여러 개의 자세(pose)가 있을 수 있으며, 첫 번째가 가장 좋은 자세/점수입니다.
        if len(suppl) > 0 and suppl[0] is not None:
            mol = suppl[0]
            # 'minimizedAffinity' 속성에 도킹 점수가 저장되어 있음
            if mol.HasProp('minimizedAffinity'):
                best_score = float(mol.GetProp('minimizedAffinity'))
        
        if best_score is not None:
            results.append({'molecule_id': molecule_id, 'docking_score': best_score})

    except Exception as e:
        print(f"{sdf_file} 처리 중 오류 발생: {e}")

# --- 3. 최종 결과 저장 ---
print("\n--- 3. 최종 결과 저장 ---")
if results:
    docking_results_df = pd.DataFrame(results)
    docking_results_df['original_index'] = docking_results_df['molecule_id'].str.replace('mol_', '').astype(int)

    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    df_with_docking = pd.merge(df, docking_results_df, left_index=True, right_on='original_index')
    
    # 불필요한 컬럼 정리
    df_with_docking.drop(columns=['molecule_id', 'original_index'], inplace=True)

    final_output_path = os.path.join(final_data_path, 'ask1_data_with_docking_scores.csv')
    df_with_docking.to_csv(final_output_path, index=False)

    print(f"\n작업 완료. 최종 결과가 '{final_output_path}'에 저장되었습니다.")
    print(df_with_docking[['smiles', 'pIC50', 'docking_score']].head())
else:
    print("처리된 도킹 결과가 없습니다. 'docking/outputs/' 폴더를 확인해주세요.")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors, Descriptors, Lipinski
from rdkit.Chem.Scaffolds import MurckoScaffold
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import optuna
import pickle
import os
import time
from datetime import datetime
import warnings
from rdkit import RDLogger

# 경고 메시지 숨기기
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')
optuna.logging.set_verbosity(optuna.logging.WARNING)

# 필요한 디렉터리 생성
os.makedirs('../outputs/models/', exist_ok=True)
os.makedirs('../outputs/figures/', exist_ok=True)

# WandB는 선택적으로 사용
try:
    import wandb
    WANDB_AVAILABLE = True
    print("WandB 사용 가능")
except ImportError:
    print("WandB가 설치되지 않았습니다. 로깅 없이 계속 진행합니다.")
    WANDB_AVAILABLE = False

print("=== Optuna를 활용한 LightGBM 1000번 최적화 실험 (Extended Features) ===")
print(f"실험 시작 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# --- 1. 최종 데이터 및 모든 피처 준비 ---
print("\n--- 1. 도킹 점수가 포함된 최종 데이터 로딩 및 피처 생성 ---")

# 파일 존재 여부 확인
try:
    df = pd.read_csv('../data/processed/ask1_data_with_docking_scores.csv')
    print(f"도킹 점수 포함 데이터 로딩 완료: {len(df)} 행")
    HAS_DOCKING = True
except FileNotFoundError:
    print("도킹 점수가 포함된 파일이 없습니다. 기본 데이터를 사용합니다.")
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"기본 데이터 로딩 완료: {len(df)} 행")
    HAS_DOCKING = False
    
    # 도킹 점수가 없는 경우 더미 점수 생성
    print("⚠️ 도킹 점수가 없어서 더미 점수를 생성합니다.")
    np.random.seed(42)
    df['docking_score'] = np.random.normal(-7.0, 1.5, len(df))

print("데이터 미리보기:")
print(df[['smiles', 'pIC50', 'docking_score']].head())

# 1a. ECFP 핑거프린트 생성
print("\n--- ECFP 핑거프린트 생성 중... ---")
mols = [Chem.MolFromSmiles(s) for s in df['smiles']]
valid_mols = [mol for mol in mols if mol is not None]
print(f"유효한 분자: {len(valid_mols)}/{len(mols)} 개")

fp = []
valid_indices = []
for i, mol in enumerate(mols):
    if mol is not None:
        try:
            fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
            fp.append(fingerprint)
            valid_indices.append(i)
        except Exception as e:
            print(f"분자 {i} 핑거프린트 생성 실패: {e}")

X_fp = np.array(fp)
print(f"ECFP 핑거프린트 생성 완료: {X_fp.shape}")

# 유효한 인덱스로 데이터프레임 필터링
df_valid = df.iloc[valid_indices].reset_index(drop=True)
mols_valid = [mols[i] for i in valid_indices]

# 1b. 확장된 물리화학적 기술자 생성
print("\n--- 물리화학적 기술자 생성 중... ---")
descriptor_names = {
    "MolWt": Descriptors.MolWt,
    "LogP": Descriptors.MolLogP,
    "TPSA": Descriptors.TPSA,
    "HBD": Lipinski.NumHDonors,
    "HBA": Lipinski.NumHAcceptors, 
    "RotatableBonds": Lipinski.NumRotatableBonds,
    "AromaticRings": Lipinski.NumAromaticRings,
    "FractionCSP3": Lipinski.FractionCSP3,
    "FormalCharge": Chem.rdmolops.GetFormalCharge,
    "RingCount": Lipinski.RingCount
}

desc_data = []
for mol in mols_valid:
    try:
        descriptors = {name: func(mol) for name, func in descriptor_names.items()}
        desc_data.append(descriptors)
    except Exception as e:
        print(f"기술자 계산 실패: {e}")
        desc_data.append({name: 0 for name in descriptor_names.keys()})

desc_df = pd.DataFrame(desc_data)
desc_df.fillna(0, inplace=True)
X_desc = desc_df.values
print(f"물리화학적 기술자 생성 완료: {X_desc.shape}")

# 1c. 도킹 점수 추출
X_docking = df_valid[['docking_score']].values
print(f"도킹 점수 추출 완료: {X_docking.shape}")

# 1d. 모든 피처셋 결합
X_combined = np.hstack((X_fp, X_desc, X_docking))
y = df_valid['pIC50'].values

print(f"\n✅ 모든 피처 준비 완료:")
print(f"   - ECFP 핑거프린트: {X_fp.shape[1]} 개")
print(f"   - 물리화학적 기술자: {X_desc.shape[1]} 개") 
print(f"   - 도킹 점수: {X_docking.shape[1]} 개")
print(f"   - 총 피처 수: {X_combined.shape[1]} 개")
print(f"   - 샘플 수: {X_combined.shape[0]} 개")

# --- 2. Scaffold 생성 ---
print("\n--- 2. Scaffold 분할 생성 ---")
scaffolds = {}
for i, mol in enumerate(mols_valid):
    try:
        scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
        if scaffold not in scaffolds:
            scaffolds[scaffold] = []
        scaffolds[scaffold].append(i)
    except Exception as e:
        # 실패한 경우 기본 그룹에 추가
        if 'failed' not in scaffolds:
            scaffolds['failed'] = []
        scaffolds['failed'].append(i)

scaffold_lists = sorted(scaffolds.values(), key=len, reverse=True)
print(f"Scaffold 그룹 생성 완료: {len(scaffold_lists)} 개")
print(f"가장 큰 그룹 크기: {len(scaffold_lists[0])} 개")

# --- 3. 향상된 Optuna 목적 함수 정의 (1000번용) ---
def enhanced_objective(trial):
    """1000번 최적화를 위한 향상된 LightGBM 하이퍼파라미터 목적 함수"""
    
    try:
        # 더 광범위한 하이퍼파라미터 탐색 공간 정의
        params = {
            'objective': 'regression',
            'metric': 'rmse',
            'boosting_type': trial.suggest_categorical('boosting_type', ['gbdt', 'dart', 'goss']),
            'random_state': 42,
            'verbosity': -1,
            'force_col_wise': True,  # 성능 최적화
            
            # 핵심 하이퍼파라미터 (더 넓은 범위)
            'num_leaves': trial.suggest_int('num_leaves', 15, 500),
            'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.5, log=True),
            'feature_fraction': trial.suggest_float('feature_fraction', 0.3, 1.0),
            'bagging_fraction': trial.suggest_float('bagging_fraction', 0.3, 1.0),
            'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
            'min_child_samples': trial.suggest_int('min_child_samples', 3, 200),
            'min_child_weight': trial.suggest_float('min_child_weight', 1e-5, 1e2, log=True),
            
            # 정규화 파라미터 (더 세밀한 조정)
            'reg_alpha': trial.suggest_float('reg_alpha', 1e-10, 100.0, log=True),
            'reg_lambda': trial.suggest_float('reg_lambda', 1e-10, 100.0, log=True),
            
            # 트리 구조 파라미터
            'max_depth': trial.suggest_int('max_depth', 3, 20),
            'min_split_gain': trial.suggest_float('min_split_gain', 0.0, 5.0),
            'max_bin': trial.suggest_int('max_bin', 100, 500),
            
            # 고급 파라미터
            'path_smooth': trial.suggest_float('path_smooth', 0.0, 200.0),
            'extra_trees': trial.suggest_categorical('extra_trees', [True, False]),
            'linear_tree': trial.suggest_categorical('linear_tree', [True, False]),
            'cat_smooth': trial.suggest_float('cat_smooth', 1.0, 100.0),
            'cat_l2': trial.suggest_float('cat_l2', 1.0, 100.0),
        }
        
        # DART 전용 파라미터
        if params['boosting_type'] == 'dart':
            params.update({
                'drop_rate': trial.suggest_float('drop_rate', 0.01, 0.5),
                'skip_drop': trial.suggest_float('skip_drop', 0.1, 0.9),
                'max_drop': trial.suggest_int('max_drop', 1, 50),
                'uniform_drop': trial.suggest_categorical('uniform_drop', [True, False])
            })
        
        # GOSS 전용 파라미터  
        if params['boosting_type'] == 'goss':
            params.update({
                'top_rate': trial.suggest_float('top_rate', 0.1, 0.5),
                'other_rate': trial.suggest_float('other_rate', 0.05, 0.2)
            })
        
        # 추정기 수 (더 넓은 범위)
        n_estimators = trial.suggest_int('n_estimators', 50, 2000)
        early_stopping_rounds = trial.suggest_int('early_stopping_rounds', 20, 100)
        
        # 3-fold 교차 검증으로 빠른 평가
        kf = KFold(n_splits=5, shuffle=True, random_state=42)
        fold_rmses = []
        
        for fold, (train_scaffold_idx, val_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
            # 인덱스 추출
            train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
            val_idx = [idx for i in val_scaffold_idx for idx in scaffold_lists[i]]
            
            if len(train_idx) < 100 or len(val_idx) < 20:  # 더 엄격한 최소 샘플 수
                continue
            
            X_train, y_train = X_combined[train_idx], y[train_idx]
            X_val, y_val = X_combined[val_idx], y[val_idx]
            
            # 모델 훈련
            model = lgb.LGBMRegressor(**params, n_estimators=n_estimators)
            model.fit(
                X_train, y_train,
                eval_set=[(X_val, y_val)],
                callbacks=[
                    lgb.early_stopping(early_stopping_rounds), 
                    lgb.log_evaluation(0)
                ]
            )
            
            # 예측 및 평가
            y_pred = model.predict(X_val)
            rmse = np.sqrt(mean_squared_error(y_val, y_pred))
            
            # 유효한 값인지 확인
            if np.isfinite(rmse) and rmse > 0:
                fold_rmses.append(rmse)
            
            # 중간 결과 보고 (pruning을 위해)
            if fold_rmses:
                trial.report(np.mean(fold_rmses), fold)
                if trial.should_prune():
                    raise optuna.exceptions.TrialPruned()
        
        if len(fold_rmses) == 0:
            return float('inf')
        
        mean_rmse = np.mean(fold_rmses)
        
        # 유효한 값 확인
        if not np.isfinite(mean_rmse) or mean_rmse <= 0:
            return float('inf')
        
        # 진행상황 출력 (100번마다)
        if trial.number % 100 == 0:
            print(f"Trial {trial.number}: RMSE={mean_rmse:.4f}")
        
        return mean_rmse
    
    except Exception as e:
        if trial.number % 100 == 0:
            print(f"Trial {trial.number} 실패: {e}")
        return float('inf')

# --- 4. Optuna 실행 (1000번 시행) ---
print(f"\n--- 4. Optuna 최적화 시작 (1000번 시행) ---")
print(f"최적화 시작 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"예상 소요 시간: 약 3-6시간 (데이터 크기에 따라)")

# Study 생성 (향상된 pruning)
study = optuna.create_study(
    direction='minimize',
    pruner=optuna.pruners.MedianPruner(
        n_startup_trials=20,  # 더 많은 초기 trial 허용
        n_warmup_steps=5,     # 더 많은 warmup
        interval_steps=1
    ),
    study_name=f"LightGBM_Extended_1000Trials_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
)

# WandB 초기화 (사용 가능한 경우)
if WANDB_AVAILABLE:
    try:
        wandb.init(
            project="ask1_lgbm_optuna_1000trials", 
            name=f"lgbm-extended-1000trials-{datetime.now().strftime('%m%d_%H%M')}",
            config={
                "total_trials": 1000,
                "features": X_combined.shape[1],
                "samples": X_combined.shape[0],
                "has_docking": HAS_DOCKING
            }
        )
        print("✅ WandB 초기화 완료")
    except Exception as e:
        print(f"⚠️ WandB 초기화 실패: {e}")
        WANDB_AVAILABLE = False

# 진행상황 추적을 위한 변수들
start_time = time.time()
best_value_history = []
trial_times = []

# 1000번 시행 실행
try:
    for i in range(1000):
        trial_start = time.time()
        
        # 100번마다 진행상황 출력
        if i % 100 == 0 and i > 0:
            elapsed = time.time() - start_time
            avg_trial_time = np.mean(trial_times[-100:]) if trial_times else elapsed / i
            remaining_trials = 1000 - i
            eta = remaining_trials * avg_trial_time / 3600  # 시간 단위
            
            print(f"\n📊 진행상황: {i}/1000 ({i/10:.1f}%)")
            print(f"   경과 시간: {elapsed/3600:.1f}시간")
            print(f"   예상 남은 시간: {eta:.1f}시간")
            if study.best_value:
                print(f"   현재 최고 RMSE: {study.best_value:.4f}")
                best_value_history.append(study.best_value)
        
        # Optuna 최적화 (1번씩)
        study.optimize(enhanced_objective, n_trials=1, timeout=600)  # 10분 타임아웃
        
        trial_time = time.time() - trial_start
        trial_times.append(trial_time)
        
        # WandB 로깅 (사용 가능한 경우, 10번마다)
        if WANDB_AVAILABLE and i % 10 == 0:
            if study.best_trial and study.best_value != float('inf'):
                wandb.log({
                    "trial": i,
                    "best_rmse": study.best_value,
                    "trials_completed": len([t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE]),
                    "trials_pruned": len([t for t in study.trials if t.state == optuna.trial.TrialState.PRUNED]),
                    "avg_trial_time": np.mean(trial_times[-10:]) if trial_times else 0
                })

except KeyboardInterrupt:
    print(f"\n사용자에 의해 중단되었습니다. (완료된 trials: {len(study.trials)})")
except Exception as e:
    print(f"최적화 중 오류 발생: {e}")

# --- 5. 결과 분석 및 정리 ---
total_time = time.time() - start_time
print(f"\n--- 5. 1000번 시행 결과 분석 ---")
print(f"완료 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"총 소요 시간: {total_time/3600:.2f} 시간")
print(f"총 시행 횟수: {len(study.trials)}")

completed_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE]
pruned_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.PRUNED]
failed_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.FAIL]

print(f"완료된 시행: {len(completed_trials)}개")
print(f"조기 종료된 시행: {len(pruned_trials)}개")
print(f"실패한 시행: {len(failed_trials)}개")
print(f"평균 trial 시간: {np.mean(trial_times):.2f}초")

if study.best_trial and len(completed_trials) > 0:
    print(f"\n🎯 1000번 최적화 결과:")
    print(f"   최적 RMSE: {study.best_value:.4f}")
    print(f"   최적 trial 번호: {study.best_trial.number}")
    print(f"   최적 하이퍼파라미터 (상위 10개):")
    
    # 주요 파라미터만 출력
    key_params = ['boosting_type', 'num_leaves', 'learning_rate', 'feature_fraction', 
                  'bagging_fraction', 'min_child_samples', 'reg_alpha', 'reg_lambda', 'max_depth']
    for key in key_params:
        if key in study.best_params:
            print(f"     {key}: {study.best_params[key]}")
    
    # WandB에 최적 결과 로깅
    if WANDB_AVAILABLE:
        wandb.log({
            "final_best_rmse": study.best_value,
            "final_best_trial": study.best_trial.number,
            "total_trials_completed": len(completed_trials),
            "total_time_hours": total_time/3600,
            "trials_per_hour": len(study.trials) / (total_time/3600)
        })

    # --- 6. 최적 모델로 최종 5-fold 평가 ---
    print(f"\n--- 6. 최적 모델로 최종 5-fold 교차 검증 ---")
    
    best_params = study.best_params.copy()
    n_estimators = best_params.pop('n_estimators', 500)
    early_stopping_rounds = best_params.pop('early_stopping_rounds', 50)
    
    # 최적 파라미터에 기본 설정 추가
    best_params.update({
        'objective': 'regression',
        'metric': 'rmse',
        'random_state': 42,
        'verbosity': -1,
        'force_col_wise': True
    })
    
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    all_preds, all_trues, fold_scores = [], [], []
    
    for fold, (train_scaffold_idx, test_scaffold_idx) in enumerate(kf.split(scaffold_lists)):
        print(f"최종 평가 Fold {fold+1}/5")
        
        train_idx = [idx for i in train_scaffold_idx for idx in scaffold_lists[i]]
        test_idx = [idx for i in test_scaffold_idx for idx in scaffold_lists[i]]
        
        X_train, y_train = X_combined[train_idx], y[train_idx]
        X_test, y_test = X_combined[test_idx], y[test_idx]
        
        # 최적 모델 훈련
        model = lgb.LGBMRegressor(**best_params, n_estimators=n_estimators)
        model.fit(X_train, y_train)
        
        # 예측 및 평가
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        
        fold_scores.append({'Fold': fold+1, 'RMSE': rmse, 'R2': r2})
        all_trues.extend(y_test)
        all_preds.extend(y_pred)
        print(f"  Fold {fold+1} RMSE: {rmse:.4f}, R²: {r2:.4f}")
    
    # 최종 성능 집계
    scores_df = pd.DataFrame(fold_scores)
    avg_rmse = scores_df['RMSE'].mean()
    std_rmse = scores_df['RMSE'].std()
    avg_r2 = scores_df['R2'].mean()
    std_r2 = scores_df['R2'].std()
    
    print(f"\n🎉 1000번 최적화된 LightGBM 최종 성능:")
    print(scores_df)
    print(f"\n📊 전체 성능:")
    print(f"   평균 RMSE: {avg_rmse:.4f} ± {std_rmse:.4f}")
    print(f"   평균 R²:   {avg_r2:.4f} ± {std_r2:.4f}")
    print(f"   총 최적화 시간: {total_time/3600:.2f} 시간")
    print(f"   시간당 trial 수: {len(study.trials)/(total_time/3600):.1f} trials/hour")
    
    # WandB에 최종 결과 로깅
    if WANDB_AVAILABLE:
        wandb.log({
            "final_avg_rmse": avg_rmse,
            "final_std_rmse": std_rmse,
            "final_avg_r2": avg_r2,
            "final_std_r2": std_r2
        })

    # --- 7. 상세한 시각화 (1000번 결과용) ---
    print(f"\n--- 7. 1000번 최적화 결과 시각화 ---")
    
    # 유효한 trial values만 필터링
    valid_trials = [t for t in completed_trials if np.isfinite(t.value) and t.value > 0]
    
    if len(valid_trials) >= 10:
        trial_numbers = [t.number for t in valid_trials]
        trial_values = [t.value for t in valid_trials]
        
        fig, axes = plt.subplots(2, 4, figsize=(20, 10))
        
        # 1. 실제 vs 예측값
        axes[0, 0].scatter(all_trues, all_preds, alpha=0.6, s=15)
        axes[0, 0].plot([min(all_trues), max(all_trues)], [min(all_trues), max(all_trues)], 'r--', linewidth=2)
        axes[0, 0].set_xlabel('Actual pIC50')
        axes[0, 0].set_ylabel('Predicted pIC50')
        axes[0, 0].set_title(f'1000-Trial Optimized Model\n(R² = {avg_r2:.3f})')
        axes[0, 0].grid(True, alpha=0.3)
        
        # 2. 최적화 히스토리
        axes[0, 1].plot(trial_numbers, trial_values, 'b.', alpha=0.7, markersize=3)
        # 이동 평균선 추가
        if len(trial_values) > 50:
            window = min(50, len(trial_values)//10)
            moving_avg = np.convolve(trial_values, np.ones(window)/window, mode='valid')
            axes[0, 1].plot(trial_numbers[window-1:], moving_avg, 'r-', linewidth=2, label=f'{window}-trial Moving Average')
            axes[0, 1].legend()
        axes[0, 1].axhline(y=study.best_value, color='green', linestyle='--', linewidth=2, label=f'Best = {study.best_value:.3f}')
        axes[0, 1].set_xlabel('Trial Number')
        axes[0, 1].set_ylabel('RMSE')
        axes[0, 1].set_title(f'1000 Trials Optimization History')
        axes[0, 1].grid(True, alpha=0.3)
        
        # 3. RMSE 분포
        axes[0, 2].hist(trial_values, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
        axes[0, 2].axvline(x=study.best_value, color='red', linestyle='--', linewidth=2, label=f'Best = {study.best_value:.3f}')
        axes[0, 2].axvline(x=np.median(trial_values), color='orange', linestyle='--', linewidth=2, label=f'Median = {np.median(trial_values):.3f}')
        axes[0, 2].set_xlabel('RMSE')
        axes[0, 2].set_ylabel('Frequency')
        axes[0, 2].set_title(f'RMSE Distribution ({len(valid_trials)} trials)')
        axes[0, 2].legend()
        axes[0, 2].grid(True, alpha=0.3)
        
        # 4. Fold별 성능
        x_pos = np.arange(len(fold_scores))
        axes[0, 3].bar(x_pos, scores_df['R2'], alpha=0.7, color='lightgreen', edgecolor='black')
        axes[0, 3].axhline(y=avg_r2, color='red', linestyle='--', linewidth=2, label=f'Mean = {avg_r2:.3f}')
        axes[0, 3].set_xlabel('Fold')
        axes[0, 3].set_ylabel('R² Score')
        axes[0, 3].set_title('Final Model Performance by Fold')
        axes[0, 3].set_xticks(x_pos)
        axes[0, 3].set_xticklabels([f'Fold {i+1}' for i in range(len(fold_scores))])
        axes[0, 3].legend()
        axes[0, 3].grid(True, alpha=0.3)
        
        # 5. 잔차 플롯
        residuals = np.array(all_trues) - np.array(all_preds)
        axes[1, 0].scatter(all_preds, residuals, alpha=0.6, s=15, c='purple')
        axes[1, 0].axhline(y=0, color='red', linestyle='--', linewidth=2)
        axes[1, 0].set_xlabel('Predicted pIC50')
        axes[1, 0].set_ylabel('Residuals')
        axes[1, 0].set_title('Residuals Plot')
        axes[1, 0].grid(True, alpha=0.3)
        
        # 6. Trial 상태 분포
        trial_states = ['Completed', 'Pruned', 'Failed']
        trial_counts = [len(completed_trials), len(pruned_trials), len(failed_trials)]
        axes[1, 1].pie(trial_counts, labels=trial_states, autopct='%1.1f%%', startangle=90)
        axes[1, 1].set_title(f'Trial Status Distribution\n(Total: {len(study.trials)})')
        
        # 7. 성능 개선 곡선 (상위 N% 기준)
        if len(valid_trials) > 100:
            percentiles = [1, 5, 10, 25, 50]
            percentile_values = [np.percentile(trial_values, p) for p in percentiles]
            axes[1, 2].plot(percentiles, percentile_values, 'o-', linewidth=2, markersize=8)
            for i, (p, v) in enumerate(zip(percentiles, percentile_values)):
                axes[1, 2].annotate(f'{v:.3f}', (p, v), textcoords="offset points", xytext=(0,10), ha='center')
            axes[1, 2].set_xlabel('Percentile')
            axes[1, 2].set_ylabel('RMSE')
            axes[1, 2].set_title('Performance Percentiles')
            axes[1, 2].grid(True, alpha=0.3)
        
        # 8. 최적화 효율성 (시간 vs 성능)
        if trial_times and len(trial_times) == len(trial_values):
            axes[1, 3].scatter(trial_times, trial_values, alpha=0.6, s=15)
            axes[1, 3].set_xlabel('Trial Time (seconds)')
            axes[1, 3].set_ylabel('RMSE')
            axes[1, 3].set_title('Optimization Efficiency\n(Time vs Performance)')
            axes[1, 3].grid(True, alpha=0.3)
        else:
            # 대신 하이퍼파라미터 중요도 표시
            axes[1, 3].text(0.5, 0.5, f'1000 Trials Completed!\n\n'
                                      f'Best RMSE: {study.best_value:.4f}\n'
                                      f'Total Time: {total_time/3600:.2f}h\n'
                                      f'Avg Trial: {np.mean(trial_times):.1f}s\n'
                                      f'Trials/Hour: {len(study.trials)/(total_time/3600):.1f}',
                            ha='center', va='center', transform=axes[1, 3].transAxes,
                            fontsize=12, bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
            axes[1, 3].set_title('1000-Trial Summary')
        
        plt.tight_layout()
        
        # 그래프 저장
        plot_path = f'../outputs/figures/optuna_lgbm_1000trials_{datetime.now().strftime("%Y%m%d_%H%M")}.png'
        try:
            plt.savefig(plot_path, dpi=300, bbox_inches='tight')
            print(f"📈 1000번 최적화 분석 플롯이 '{plot_path}'에 저장되었습니다.")
            
            if WANDB_AVAILABLE:
                wandb.log({"1000trials_analysis": wandb.Image(plot_path)})
        except Exception as e:
            print(f"⚠️ 그래프 저장 실패: {e}")
        
        plt.show()

# --- 8. Study 및 결과 저장 ---
study_path = f'../outputs/models/lgbm_extended_1000trials_study_{datetime.now().strftime("%Y%m%d_%H%M%S")}.pkl'
try:
    with open(study_path, 'wb') as f:
        pickle.dump(study, f)
    print(f"📁 1000번 최적화 study가 {study_path}에 저장되었습니다.")
except Exception as e:
    print(f"⚠️ Study 저장 실패: {e}")

# 상위 20개 Trial 분석 및 저장
if len(completed_trials) >= 20:
    print(f"\n--- 8. 상위 20개 Trial 상세 분석 ---")
    
    top_20_trials = sorted(completed_trials, key=lambda t: t.value)[:20]
    
    print("🏆 Top 20 Trials:")
    for i, trial in enumerate(top_20_trials, 1):
        print(f"   {i:2d}. Trial {trial.number}: RMSE = {trial.value:.4f}")
    
    # 상위 20개의 하이퍼파라미터 패턴 분석
    top_params_analysis = {}
    for param_name in top_20_trials[0].params.keys():
        param_values = [trial.params[param_name] for trial in top_20_trials]
        
        if isinstance(param_values[0], (int, float)):
            top_params_analysis[param_name] = {
                'mean': np.mean(param_values),
                'std': np.std(param_values),
                'min': np.min(param_values),
                'max': np.max(param_values),
                'median': np.median(param_values)
            }
        else:
            from collections import Counter
            counter = Counter(param_values)
            total = len(param_values)
            top_params_analysis[param_name] = {
                'most_common': counter.most_common(3),
                'total_variants': len(counter)
            }
    
    print(f"\n📊 Top 20 Trial 하이퍼파라미터 패턴:")
    for param, stats in top_params_analysis.items():
        if isinstance(stats, dict) and 'mean' in stats:
            print(f"   {param}: {stats['mean']:.3f} ± {stats['std']:.3f} (범위: {stats['min']:.3f}-{stats['max']:.3f})")
        else:
            print(f"   {param}: {stats['most_common']}")
    
    # Top 20 결과를 CSV로 저장
    top_20_data = []
    for trial in top_20_trials:
        row = {'trial_number': trial.number, 'rmse': trial.value}
        row.update(trial.params)
        top_20_data.append(row)
    
    top_20_df = pd.DataFrame(top_20_data)
    top_20_path = f'../outputs/models/top_20_trials_1000_{datetime.now().strftime("%Y%m%d_%H%M")}.csv'
    top_20_df.to_csv(top_20_path, index=False)
    print(f"📊 Top 20 trials 상세 정보가 {top_20_path}에 저장되었습니다.")

# WandB 종료
if WANDB_AVAILABLE:
    try:
        wandb.finish()
        print("✅ WandB 세션 정상 종료")
    except Exception as e:
        print(f"⚠️ WandB 종료 중 오류: {e}")

print(f"\n🎉 1000번 Optuna 최적화 실험 완료!")
print(f"종료 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"총 소요 시간: {total_time/3600:.2f} 시간")
print(f"최종 최적 RMSE: {study.best_value:.4f}" if study.best_trial else "최적 결과 없음")
print(f"결과 파일들이 '../outputs/' 디렉터리에 저장되었습니다.")
print(f"\n🔥 1000번의 시행으로 최고의 하이퍼파라미터를 찾았습니다!")
```

```python
import os
import time
import pickle
import warnings
from datetime import datetime

import numpy as np
import pandas as pd
import xgboost as xgb
from rdkit import Chem, RDLogger, DataStructs
from rdkit.Chem import rdMolDescriptors, Descriptors, Lipinski
from rdkit.Chem.Scaffolds import MurckoScaffold
from sklearn.model_selection import GroupKFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import optuna

# =========================
# 기본 설정
# =========================
warnings.filterwarnings('ignore')
RDLogger.DisableLog('rdApp.*')
optuna.logging.set_verbosity(optuna.logging.WARNING)

os.makedirs('../outputs/models/', exist_ok=True)
os.makedirs('../outputs/figures/', exist_ok=True)

# WandB는 선택적으로 사용
try:
    import wandb
    WANDB_AVAILABLE = True
    print("WandB 사용 가능")
except ImportError:
    print("WandB가 설치되지 않았습니다. 로깅 없이 계속 진행합니다.")
    WANDB_AVAILABLE = False

print("=== Optuna를 활용한 XGBoost 100번 최적화 실험 (DMatrix EarlyStopping, GroupKFold, GPU) ===")
print(f"실험 시작 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"XGBoost 버전: {xgb.__version__}")

# =========================
# 유틸
# =========================
def get_xgb_version_tuple():
    v = xgb.__version__.split('.')[:2]
    return tuple(map(int, v))

XGB_VERSION = get_xgb_version_tuple()
print(f"XGBoost 버전(major, minor): {XGB_VERSION}")

def make_group_kfold(n_splits: int, groups: np.ndarray) -> GroupKFold:
    unique = np.unique(groups)
    k = max(2, min(n_splits, len(unique)))
    return GroupKFold(n_splits=k)

def convert_bitvecs_to_numpy(bitvec_list, n_bits=2048) -> np.ndarray:
    X = np.zeros((len(bitvec_list), n_bits), dtype=np.float32)
    tmp = np.zeros((n_bits,), dtype=np.int32)
    for i, bv in enumerate(bitvec_list):
        tmp[:] = 0
        DataStructs.ConvertToNumpyArray(bv, tmp)
        X[i] = tmp
    return X

def booster_predict(booster: xgb.Booster, X: np.ndarray) -> np.ndarray:
    dm = xgb.DMatrix(X)
    ntree_limit = getattr(booster, "best_ntree_limit", None)
    if ntree_limit is not None and ntree_limit > 0:
        try:
            return booster.predict(dm, ntree_limit=ntree_limit)
        except TypeError:
            pass
    best_iteration = getattr(booster, "best_iteration", None)
    if best_iteration is not None:
        try:
            return booster.predict(dm, iteration_range=(0, best_iteration + 1))
        except TypeError:
            pass
    return booster.predict(dm)

def apply_gpu_params(params: dict) -> dict:
    """XGBoost 버전에 맞춰 GPU 파라미터 적용"""
    p = dict(params)
    if XGB_VERSION >= (2, 0):
        p['device'] = 'cuda'
        p['tree_method'] = 'hist'
        # 2.x에서는 predictor 자동
    else:
        p['tree_method'] = 'gpu_hist'
        p['predictor'] = 'gpu_predictor'
    return p

def train_with_early_stopping(params: dict,
                              X_train: np.ndarray, y_train: np.ndarray,
                              X_val: np.ndarray, y_val: np.ndarray,
                              num_boost_round: int,
                              early_stopping_rounds: int) -> xgb.Booster:
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)
    watchlist = [(dtrain, 'train'), (dval, 'eval')]
    booster = xgb.train(
        params=params,
        dtrain=dtrain,
        num_boost_round=num_boost_round,
        evals=watchlist,
        early_stopping_rounds=early_stopping_rounds,
        verbose_eval=False
    )
    return booster

# =========================
# 1. 데이터 로딩 & 피처 생성
# =========================
print("\n--- 1. 도킹 점수가 포함된 최종 데이터 로딩 및 피처 생성 ---")
try:
    df = pd.read_csv('../data/processed/ask1_data_with_docking_scores.csv')
    print(f"도킹 점수 포함 데이터 로딩 완료: {len(df)} 행")
    HAS_DOCKING = True
except FileNotFoundError:
    print("도킹 점수가 포함된 파일이 없습니다. 기본 데이터를 사용합니다.")
    df = pd.read_csv('../data/processed/ask1_master_final.csv')
    print(f"기본 데이터 로딩 완료: {len(df)} 행")
    HAS_DOCKING = False
    print("⚠️ 도킹 점수가 없어서 더미 점수를 생성합니다.")
    np.random.seed(42)
    df['docking_score'] = np.random.normal(-7.0, 1.5, len(df))

print("데이터 미리보기:")
print(df[['smiles', 'pIC50', 'docking_score']].head())

# 1a. SMILES -> Mol
mols_all = [Chem.MolFromSmiles(s) for s in df['smiles']]
valid_indices = [i for i, m in enumerate(mols_all) if m is not None]
mols = [mols_all[i] for i in valid_indices]
df_valid = df.iloc[valid_indices].reset_index(drop=True)
print(f"유효한 분자: {len(mols)}/{len(mols_all)} 개")

# 1b. ECFP (Morgan r=2, 2048bits)
print("\n--- ECFP 핑거프린트 생성 중... ---")
n_bits = 2048
ecfp_bitvecs = []
for i, mol in enumerate(mols):
    try:
        bv = rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=n_bits)
        ecfp_bitvecs.append(bv)
    except Exception as e:
        print(f"분자 {i} ECFP 생성 실패: {e}")
        ecfp_bitvecs.append(DataStructs.ExplicitBitVect(n_bits))
X_fp = convert_bitvecs_to_numpy(ecfp_bitvecs, n_bits=n_bits)
print(f"ECFP 핑거프린트 생성 완료: {X_fp.shape}")

# 1c. 확장 물리화학적 기술자
print("\n--- 물리화학적 기술자 생성 중... ---")
descriptor_fns = {
    "MolWt": Descriptors.MolWt,
    "LogP": Descriptors.MolLogP,
    "TPSA": Descriptors.TPSA,
    "HBD": Lipinski.NumHDonors,
    "HBA": Lipinski.NumHAcceptors,
    "RotatableBonds": Lipinski.NumRotatableBonds,
    "AromaticRings": Lipinski.NumAromaticRings,
    "FractionCSP3": Lipinski.FractionCSP3,
    "FormalCharge": Chem.rdmolops.GetFormalCharge,
    "RingCount": Lipinski.RingCount
}
desc_rows = []
for mol in mols:
    try:
        desc_rows.append({k: fn(mol) for k, fn in descriptor_fns.items()})
    except Exception as e:
        print(f"기술자 계산 실패: {e}")
        desc_rows.append({k: 0 for k in descriptor_fns})
desc_df = pd.DataFrame(desc_rows).fillna(0)
X_desc = desc_df.astype(np.float32).values
print(f"물리화학적 기술자 생성 완료: {X_desc.shape}")

# 1d. 도킹 점수
X_docking = df_valid[['docking_score']].astype(np.float32).values
print(f"도킹 점수 추출 완료: {X_docking.shape}")

# 1e. 타깃
y = df_valid['pIC50'].astype(np.float32).values

# 1f. 전체 피처 결합
X_combined = np.hstack([X_fp, X_desc, X_docking]).astype(np.float32)
print(f"\n✅ 모든 피처 준비 완료:")
print(f"   - ECFP 핑거프린트: {X_fp.shape[1]} 개")
print(f"   - 물리화학적 기술자: {X_desc.shape[1]} 개")
print(f"   - 도킹 점수: {X_docking.shape[1]} 개")
print(f"   - 총 피처 수: {X_combined.shape[1]} 개")
print(f"   - 샘플 수: {X_combined.shape[0]} 개")

# =========================
# 2. Scaffold 그룹 (groups per sample)
# =========================
print("\n--- 2. Scaffold 문자열 생성 (GroupKFold용) ---")
scaffold_strings = []
for mol in mols:
    try:
        s = MurckoScaffold.MurckoScaffoldSmiles(mol=mol, includeChirality=False)
        scaffold_strings.append(s if s else "failed")
    except Exception:
        scaffold_strings.append("failed")
scaffold_groups = np.array(scaffold_strings)
print(f"유니크 scaffold 그룹 수: {len(np.unique(scaffold_groups))}")

# =========================
# 3. Optuna 목적 함수 (xgb.train + GPU)
# =========================
def enhanced_xgboost_objective(trial):
    params = {
        'objective': 'reg:squarederror',
        'eta': trial.suggest_float('learning_rate', 0.03, 0.3, log=True),
        'min_child_weight': trial.suggest_float('min_child_weight', 1.0, 10.0),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'alpha': trial.suggest_float('reg_alpha', 1e-5, 1.0, log=True),
        'lambda': trial.suggest_float('reg_lambda', 1e-5, 1.0, log=True),
        'gamma': trial.suggest_float('gamma', 1e-5, 1.0, log=True),
        'grow_policy': trial.suggest_categorical('grow_policy', ['depthwise', 'lossguide']),
        'verbosity': 0,
        'eval_metric': 'rmse',
        'seed': 42
    }
    # GPU 적용
    params = apply_gpu_params(params)

    num_boost_round = trial.suggest_int('n_estimators', 200, 800)
    early_rounds = trial.suggest_int('early_stopping_rounds', 20, 80)

    if params['grow_policy'] == 'lossguide':
        params['max_depth'] = 0
        params['max_leaves'] = trial.suggest_int('max_leaves', 32, 256)
    else:
        params['max_depth'] = trial.suggest_int('max_depth', 3, 12)

    outer_gkf = make_group_kfold(3, scaffold_groups)
    fold_rmses = []

    for fold_id, (tr_idx, te_idx) in enumerate(outer_gkf.split(X_combined, y, groups=scaffold_groups)):
        if len(tr_idx) < 50 or len(te_idx) < 10:
            continue

        inner_groups = scaffold_groups[tr_idx]
        inner_gkf = make_group_kfold(5, inner_groups)
        try:
            inner_tr_idx, inner_val_idx = next(inner_gkf.split(X_combined[tr_idx], y[tr_idx], groups=inner_groups))
        except StopIteration:
            inner_tr_idx = np.arange(len(tr_idx))[:int(0.8 * len(tr_idx))]
            inner_val_idx = np.arange(len(tr_idx))[int(0.8 * len(tr_idx)):]

        train_idx = tr_idx[inner_tr_idx]
        val_idx = tr_idx[inner_val_idx]
        test_idx = te_idx

        booster = train_with_early_stopping(
            params=params,
            X_train=X_combined[train_idx], y_train=y[train_idx],
            X_val=X_combined[val_idx],   y_val=y[val_idx],
            num_boost_round=num_boost_round,
            early_stopping_rounds=early_rounds
        )

        y_pred = booster_predict(booster, X_combined[test_idx])
        rmse = float(np.sqrt(mean_squared_error(y[test_idx], y_pred)))
        if np.isfinite(rmse) and rmse > 0:
            fold_rmses.append(rmse)

        if fold_rmses:
            trial.report(np.mean(fold_rmses), step=fold_id)
            if trial.should_prune():
                raise optuna.exceptions.TrialPruned()

    return np.mean(fold_rmses) if fold_rmses else float('inf')

# =========================
# 4. Optuna 실행 (200 trials)
# =========================
print(f"\n--- 4. XGBoost Optuna 최적화 시작 (200 trials, GPU) ---")
print(f"최적화 시작 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

study = optuna.create_study(
    direction='minimize',
    pruner=optuna.pruners.MedianPruner(
        n_startup_trials=5,
        n_warmup_steps=1,
        interval_steps=1
    ),
    study_name=f"XGBoost_GPU_100Trials_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
)

if WANDB_AVAILABLE:
    try:
        wandb.init(
            project="ask1_xgboost_fixed",
            name=f"xgb-gpu-100trials-{datetime.now().strftime('%m%d_%H%M')}",
            config={
                "total_trials": 200,
                "features": int(X_combined.shape[1]),
                "samples": int(X_combined.shape[0]),
                "has_docking": bool('docking_score' in df_valid.columns),
                "xgboost_version": xgb.__version__,
                "gpu_mode": "device=cuda" if XGB_VERSION >= (2,0) else "gpu_hist/predictor"
            }
        )
        print("✅ WandB 초기화 완료")
    except Exception as e:
        print(f"⚠️ WandB 초기화 실패: {e}")
        WANDB_AVAILABLE = False

start_time = time.time()

try:
    study.optimize(enhanced_xgboost_objective, n_trials=100)
    if WANDB_AVAILABLE:
        for i, t in enumerate(study.trials):
            if t.state == optuna.trial.TrialState.COMPLETE and np.isfinite(t.value):
                wandb.log({"trial": i, "rmse": t.value, **{f"param/{k}": v for k, v in t.params.items()}})
except KeyboardInterrupt:
    print(f"\n사용자에 의해 중단되었습니다. (완료된 trials: {len(study.trials)})")
except Exception as e:
    print(f"최적화 중 오류 발생: {e}")

# =========================
# 5. 결과 분석
# =========================
total_time = time.time() - start_time
print(f"\n--- 5. XGBoost 100번 시행 결과 분석 ---")
print(f"완료 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"총 소요 시간: {total_time:.2f} 초")
print(f"총 시행 횟수: {len(study.trials)}")

completed_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE and np.isfinite(t.value) and t.value > 0]
pruned_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.PRUNED]
failed_trials = [t for t in study.trials if t.state == optuna.trial.TrialState.FAIL]

print(f"완료된 시행: {len(completed_trials)}개")
print(f"조기 종료된 시행: {len(pruned_trials)}개")
print(f"실패한 시행: {len(failed_trials)}개")

# =========================
# 6. 최적 파라미터로 5-fold 평가 (GPU, DMatrix)
# =========================
if completed_trials:
    best_trial = min(completed_trials, key=lambda t: t.value)
    print(f"\n🎯 최적 RMSE: {best_trial.value:.4f}")
    print(f"   최적 trial 번호: {best_trial.number}")
    print("   최적 하이퍼파라미터:")
    for k, v in best_trial.params.items():
        print(f"     {k}: {v}")

    if WANDB_AVAILABLE:
        wandb.log({
            "final_best_rmse": best_trial.value,
            "final_best_trial": best_trial.number,
            "total_trials_completed": len(completed_trials),
            "total_time_seconds": total_time,
            "xgb_version_major": XGB_VERSION[0],
            "xgb_version_minor": XGB_VERSION[1]
        })

    p = dict(best_trial.params)
    train_params = {
        'objective': 'reg:squarederror',
        'eta': p.get('learning_rate', 0.1),
        'min_child_weight': p['min_child_weight'],
        'subsample': p['subsample'],
        'colsample_bytree': p['colsample_bytree'],
        'alpha': p['reg_alpha'],
        'lambda': p['reg_lambda'],
        'gamma': p['gamma'],
        'grow_policy': p['grow_policy'],
        'verbosity': 0,
        'eval_metric': 'rmse',
        'seed': 42
    }
    if p['grow_policy'] == 'lossguide':
        train_params['max_depth'] = 0
        train_params['max_leaves'] = p.get('max_leaves', 64)
    else:
        train_params['max_depth'] = p.get('max_depth', 6)
    # GPU 적용
    train_params = apply_gpu_params(train_params)

    num_boost_round = int(p.get('n_estimators', 400))
    early_rounds = int(p.get('early_stopping_rounds', 40))

    outer_gkf = make_group_kfold(5, scaffold_groups)

    all_preds, all_trues, fold_scores = [], [], []
    for fold_id, (tr_idx, te_idx) in enumerate(outer_gkf.split(X_combined, y, groups=scaffold_groups), start=1):
        print(f"최종 평가 Fold {fold_id}")
        inner_groups = scaffold_groups[tr_idx]
        inner_gkf = make_group_kfold(5, inner_groups)
        try:
            inner_tr_idx, inner_val_idx = next(inner_gkf.split(X_combined[tr_idx], y[tr_idx], groups=inner_groups))
        except StopIteration:
            inner_tr_idx = np.arange(len(tr_idx))[:int(0.8 * len(tr_idx))]
            inner_val_idx = np.arange(len(tr_idx))[int(0.8 * len(tr_idx)):]
        train_idx = tr_idx[inner_tr_idx]
        val_idx = tr_idx[inner_val_idx]
        test_idx = te_idx

        booster = train_with_early_stopping(
            params=train_params,
            X_train=X_combined[train_idx], y_train=y[train_idx],
            X_val=X_combined[val_idx],   y_val=y[val_idx],
            num_boost_round=num_boost_round,
            early_stopping_rounds=early_rounds
        )

        y_pred = booster_predict(booster, X_combined[test_idx])
        rmse = float(np.sqrt(mean_squared_error(y[test_idx], y_pred)))
        r2 = float(r2_score(y[test_idx], y_pred))
        fold_scores.append({'Fold': fold_id, 'RMSE': rmse, 'R2': r2})
        all_trues.extend(y[test_idx].tolist())
        all_preds.extend(y_pred.tolist())
        print(f"  Fold {fold_id} RMSE: {rmse:.4f}, R²: {r2:.4f}")

    scores_df = pd.DataFrame(fold_scores)
    avg_rmse = scores_df['RMSE'].mean()
    std_rmse = scores_df['RMSE'].std()
    avg_r2 = scores_df['R2'].mean()
    std_r2 = scores_df['R2'].std()

    print(f"\n🎉 최종 성능(5-fold, scaffold-aware):")
    print(scores_df)
    print(f"\n📊 전체 성능: 평균 RMSE {avg_rmse:.4f} ± {std_rmse:.4f} / 평균 R² {avg_r2:.4f} ± {std_r2:.4f}")
    print(f"총 최적화 시간: {total_time:.2f} 초")

    if WANDB_AVAILABLE:
        wandb.log({
            "final_avg_rmse": avg_rmse,
            "final_std_rmse": std_rmse,
            "final_avg_r2": avg_r2,
            "final_std_r2": std_r2
        })

    # =========================
    # 7. 시각화
    # =========================
    print(f"\n--- 7. 결과 시각화 ---")
    if len(all_trues) > 0 and len(all_preds) > 0:
        fig, axes = plt.subplots(2, 3, figsize=(15, 10))

        axes[0, 0].scatter(all_trues, all_preds, alpha=0.6, s=18)
        lo = min(min(all_trues), min(all_preds))
        hi = max(max(all_trues), max(all_preds))
        axes[0, 0].plot([lo, hi], [lo, hi], '--', linewidth=1.5)
        axes[0, 0].set_xlabel('Actual pIC50')
        axes[0, 0].set_ylabel('Predicted pIC50')
        axes[0, 0].set_title(f'XGBoost (100 Trials, GPU)\nMean R² = {avg_r2:.3f}')
        axes[0, 0].grid(True, alpha=0.3)

        valid_trials = completed_trials
        if valid_trials:
            trial_numbers = [t.number for t in valid_trials]
            trial_values = [t.value for t in valid_trials]
            axes[0, 1].plot(trial_numbers, trial_values, 'o-', linewidth=1.5, markersize=6)
            axes[0, 1].axhline(y=best_trial.value, linestyle='--', linewidth=1.5,
                               label=f'Best = {best_trial.value:.3f}')
            axes[0, 1].legend()
        axes[0, 1].set_xlabel('Trial Number')
        axes[0, 1].set_ylabel('RMSE')
        axes[0, 1].set_title('Optuna 100 Trials')
        axes[0, 1].grid(True, alpha=0.3)

        x_pos = np.arange(len(fold_scores))
        axes[0, 2].bar(x_pos, scores_df['R2'], alpha=0.8, edgecolor='black')
        axes[0, 2].axhline(y=avg_r2, linestyle='--', linewidth=1.5, label=f'Mean = {avg_r2:.3f}')
        axes[0, 2].set_xlabel('Fold')
        axes[0, 2].set_ylabel('R² Score')
        axes[0, 2].set_title('Performance by Fold')
        axes[0, 2].set_xticks(x_pos)
        axes[0, 2].set_xticklabels([f'Fold {i+1}' for i in range(len(fold_scores))])
        axes[0, 2].legend()
        axes[0, 2].grid(True, alpha=0.3)

        residuals = np.array(all_trues) - np.array(all_preds)
        axes[1, 0].scatter(all_preds, residuals, alpha=0.6, s=18)
        axes[1, 0].axhline(y=0, linestyle='--', linewidth=1.5)
        axes[1, 0].set_xlabel('Predicted pIC50')
        axes[1, 0].set_ylabel('Residuals')
        axes[1, 0].set_title('Residuals')
        axes[1, 0].grid(True, alpha=0.3)

        trial_states = ['Completed', 'Pruned', 'Failed']
        trial_counts = [len(completed_trials), len(pruned_trials), len(failed_trials)]
        axes[1, 1].pie(trial_counts, labels=trial_states, autopct='%1.1f%%', startangle=90)
        axes[1, 1].set_title(f'Trial Status\n(Total: {len(study.trials)})')

        axes[1, 2].text(
            0.5, 0.5,
            f'XGBoost Results (GPU)\n\n'
            f'Best RMSE: {best_trial.value:.4f}\n'
            f'Mean R²: {avg_r2:.4f}\n'
            f'Total Time: {total_time:.1f}s\n'
            f'Valid Trials: {len(valid_trials)}/100\n'
            f'XGBoost v{xgb.__version__}',
            ha='center', va='center', transform=axes[1, 2].transAxes,
            fontsize=11, bbox=dict(boxstyle='round', alpha=0.8)
        )
        axes[1, 2].set_title('Summary')
        axes[1, 2].axis('off')

        plt.tight_layout()

        plot_path = f'../outputs/figures/xgboost_gpu_100trials_{datetime.now().strftime("%Y%m%d_%H%M")}.png'
        try:
            plt.savefig(plot_path, dpi=300, bbox_inches='tight')
            print(f"📈 결과 플롯 저장: {plot_path}")
            if WANDB_AVAILABLE:
                wandb.log({"xgboost_gpu_100trials": wandb.Image(plot_path)})
        except Exception as e:
            print(f"⚠️ 그래프 저장 실패: {e}")

        plt.show()

else:
    print("⚠️ 유효한 최적화 결과가 없습니다. XGBoost/데이터 설정을 확인하세요.")

# =========================
# 8. Study 저장
# =========================
study_path = f'../outputs/models/xgboost_gpu_100trials_study_{datetime.now().strftime("%Y%m%d_%H%M%S")}.pkl'
try:
    with open(study_path, 'wb') as f:
        pickle.dump(study, f)
    print(f"📁 Study 저장 완료: {study_path}")
except Exception as e:
    print(f"⚠️ Study 저장 실패: {e}")

# =========================
# 9. WandB 종료
# =========================
if WANDB_AVAILABLE:
    try:
        wandb.finish()
        print("✅ WandB 세션 정상 종료")
    except Exception as e:
        print(f"⚠️ WandB 종료 중 오류: {e}")

print(f"\n🎉 테스트 완료! 종료 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
if completed_trials:
    print(f"최종 최적 RMSE: {best_trial.value:.4f}")
    print(f"성공률: {len(completed_trials)/100*100:.1f}%")
else:
    print("최적 결과 없음 - 추가 디버깅 필요")
```
